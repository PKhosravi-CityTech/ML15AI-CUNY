# From CNNs to RNNs: Building Towards Sequential Understanding

**Convolutional Neural Networks (CNNs)** and **Recurrent Neural Networks (RNNs)** are two foundational architectures in deep learning. While CNNs specialize in capturing **spatial features** from images, RNNs are designed for **sequential data**, where the order of information matters — like in videos, time series, and language.

This session bridges the transition from CNNs to RNNs, preparing us to dive into natural language processing (NLP), where understanding sequences is essential.

## CNNs in Video and Image Analysis

CNNs excel at identifying patterns within fixed-size spatial data such as images.  
In video analysis, CNNs are commonly used to extract spatial features from individual frames, treating each frame as an image. These extracted features — such as objects, poses, or textures — form the basis for higher-level reasoning.

**Example:**  
- In action recognition, CNNs can detect the presence and position of a person or object within each frame.

## RNNs in Sequential Analysis

While CNNs capture *what* is present in a single frame, RNNs capture *how* things change over time.

Videos, speech, and text are inherently sequential: the meaning depends not only on individual frames or words, but also on their order and evolution. RNNs process sequences step-by-step, maintaining an internal hidden state that summarizes previous information — allowing the model to capture temporal dependencies.

**Applications:**
- Action recognition across multiple frames,
- Video captioning (describing a sequence of frames with text),
- Speech recognition,
- Natural language modeling.

## CNN-RNN Fusion Architectures

Modern systems often combine CNNs and RNNs:  
- **CNNs** extract spatial features frame-by-frame.  
- **RNNs** model the temporal relationships between those features across frames.

This fusion enables simultaneous spatial and temporal reasoning — crucial for understanding dynamic scenes in video or sequential text patterns in NLP.

**Example:**  
In video surveillance, CNN-RNN models can detect an object and track its movement across time to recognize suspicious behavior.

## Challenges in CNN-RNN Models

Building effective CNN-RNN models presents several challenges:
- **Computational Load:** Videos and long sequences require heavy processing and memory.
- **Long-Term Dependencies:** Basic RNNs struggle with long sequences due to the vanishing gradient problem, making it difficult to learn relationships far apart in time.
- **Hyperparameter Sensitivity:** Model depth, hidden state size, and sequence length significantly impact performance.

To address some of these limitations, more advanced RNN architectures were developed.

# Recurrent Neural Networks (RNNs) and Their Variants

Recurrent Neural Networks (RNNs) are designed to process sequential data by maintaining a hidden state that evolves over time as the input sequence is processed.

However, traditional (vanilla) RNNs suffer from:
- Vanishing gradients (struggling to learn long-term dependencies),
- Exploding gradients (unstable training).

These issues led to the development of **advanced RNN variants**:

## Long Short-Term Memory (LSTM)

LSTM networks introduce a memory cell and three gates — input, forget, and output — which control the flow of information.  
This architecture allows LSTMs to selectively remember or forget information, making them highly effective at modeling long-range dependencies in sequences.

**Key strengths:**
- Capture long-term patterns,
- Resist vanishing gradients,
- Widely used in tasks like machine translation, speech recognition, and text generation.

## Gated Recurrent Unit (GRU)

GRUs are a streamlined alternative to LSTMs.  
They use a simpler gating mechanism (update and reset gates) while achieving similar performance with fewer parameters, making them:
- Faster to train,
- Less computationally intensive.

GRUs strike a good balance between expressive power and computational efficiency, and are preferred in scenarios where training speed is critical.

## GRU and LSTM Mathematical Equations

### GRU Equations

The GRU (Gated Recurrent Unit) uses two gates: **Update Gate** and **Reset Gate**. The following equations define the full GRU update mechanism:

- **Update Gate:**
$$z_t = \sigma(W_z \cdot x_t + U_z \cdot h_{t-1})$$

- **Reset Gate:**
$$r_t = \sigma(W_r \cdot x_t + U_r \cdot h_{t-1})$$

- **Candidate Hidden State:**
$$\tilde{h}_t = \tanh(W_h \cdot x_t + U_h \cdot (r_t \odot h_{t-1}))$$

- **Final Hidden State:**
$$h_t = (1 - z_t) \odot h_{t-1} + z_t \odot \tilde{h}_t$$

These equations ensure that the GRU can control how much past information to keep or reset.

### LSTM Equations

The LSTM (Long Short-Term Memory) uses three gates: **Forget Gate**, **Input Gate**, and **Output Gate**. The update equations also include a **candidate cell state**, **cell state**, and **hidden state**:

- **Forget Gate:**
$$f_t = \sigma(W_f \cdot x_t + U_f \cdot h_{t-1})$$

- **Input Gate:**
$$i_t = \sigma(W_i \cdot x_t + U_i \cdot h_{t-1})$$

- **Output Gate:**
$$o_t = \sigma(W_o \cdot x_t + U_o \cdot h_{t-1})$$

- **Candidate Cell State:**
$$\tilde{c}_t = \tanh(W_c \cdot x_t + U_c \cdot h_{t-1})$$

- **Cell State Update:**
$$c_t = f_t \odot c_{t-1} + i_t \odot \tilde{c}_t$$

- **Hidden State:**
$$h_t = o_t \odot \tanh(c_t)$$

These equations describe how LSTM manages long-term dependencies by balancing information flow using its gates.


## Understanding Symbols in GRU and LSTM Equations

In the GRU (Gated Recurrent Unit) and LSTM (Long Short-Term Memory) update equations, various symbols are used to represent different components of the models. Here is a clear explanation of what each symbol represents:

### Symbol Definitions

| Symbol | Meaning |
|:-------|:--------|
| **$x_t$** | Input at time $t$ |
| **$h_{t-1}$** | Hidden state at previous time step $t-1$ |
| **$h_t$** | Hidden state at current time step $t$ |
| **$c_{t-1}$** | Cell state at previous time step (LSTM only) |
| **$c_t$** | Cell state at current time step (LSTM only) |
| **$W$** | Weight matrix for the **current input** $x_t$ |
| **$U$** | Weight matrix for the **previous hidden state** $h_{t-1}$ |
| **$z_t$** | Update gate (GRU) |
| **$r_t$** | Reset gate (GRU) |
| **$\tilde{h}_t$** | Candidate hidden state (GRU) |
| **$i_t$** | Input gate (LSTM) |
| **$f_t$** | Forget gate (LSTM) |
| **$o_t$** | Output gate (LSTM) |
| **$\tilde{c}_t$** | Candidate cell state (LSTM) |
| **$\sigma$** | Sigmoid activation function |
| **$\tanh$** | Hyperbolic tangent activation function |
| **$\odot$** | Element-wise (Hadamard) product |

### Summary

- **$W$** and **$U$** are learned parameters during training.
- Gate variables like $z_t$, $r_t$, $i_t$, $f_t$, and $o_t$ are dynamic activations computed at each time step.
- **$\sigma$** and **$\tanh$** apply non-linear transformations to the linear combinations of inputs and hidden states.

This structure helps GRUs and LSTMs to effectively capture dependencies across sequences, manage memory, and control the flow of information.



## Project Overview: CNN + GRU for Video Action Recognition

### Project Goal

In this notebook, we aim to explore how to model sequential data by combining Convolutional Neural Networks (CNNs) and Recurrent Neural Networks (RNNs).

Specifically, we will:
- Use a pre-trained CNN (ResNet50) to extract spatial features from individual video frames.
- Model the temporal sequence of these features using a Gated Recurrent Unit (GRU) network.
- Train the system to classify videos into one of five human actions.


### About the Dataset

We are using a small subset of the UCF101 dataset, a popular benchmark for human action recognition.

The subset (`ucf101_top5`) contains videos from five different action categories:

- Cricket Shot
- Playing Cello

Each video consists of a short clip showing a specific action performed by different individuals in diverse environments.

### Workflow Overview

Here is the pipeline we will follow:

```text
Video Frames ➞ CNN (ResNet50) ➞ Frame Features ➞ GRU ➞ Action Class
```

- **Video Frames:** Load and preprocess the video clips frame-by-frame.
- **CNN (ResNet50):** Extract high-level spatial features from each frame.
- **Frame Features:** Assemble extracted features into a sequence.
- **GRU:** Model temporal dependencies between frames.
- **Action Class:** Predict the action performed in the video.

**In this code, we perform transfer learning by feature extraction:**
We use a pre-trained CNN (ResNet50) to extract fixed features from individual video frames and train a new GRU-based sequence model on top. The CNN weights are not fine-tuned, ensuring fast training and reduced computational cost. This notebook implements a video classification pipeline using a CNN-GRU architecture on a subset of the UCF101 dataset (ucf101_top5). The workflow begins with importing necessary libraries and preparing the dataset, including metadata loading and frame preprocessing. Each video is processed into a sequence of up to 20 resized and normalized frames. A pre-trained ResNet50 model (with the final classification layer removed) acts as a feature extractor, converting each frame into a 2048-dimensional vector. These frame-level features are then input to a two-layer GRU, which captures temporal dependencies across frames. The GRU's final hidden state is passed through a dropout layer and a fully connected layer to predict one of the five action categories. The model is trained with cross-entropy loss and evaluated using accuracy on a separate test set.

````python
# PyTorch Video Classification with CNN-GRU on UCF101_top5 Dataset
# ---------------------------------------------------------------
# This notebook loads a small video dataset, extracts CNN features per frame,
# then models sequences with a GRU to classify the action happening in each video.
# The code is simplified, fast to run, and fully Colab-compatible.

# Step 1: Import Packages
import os  # For file and directory operations
import cv2  # OpenCV for video frame reading
import torch  # PyTorch main package
import imageio  # Reading and writing image data (not heavily used here)
import random  # For random number generation
import numpy as np  # Numerical operations
import pandas as pd  # For data manipulation and CSV reading
import matplotlib.pyplot as plt  # For plotting

from torch import nn  # Neural network components
from torchvision import models, transforms  # Pretrained models and image transforms
from torch.utils.data import Dataset, DataLoader  # Dataset and data loading tools
from torchvision.models import resnet50, ResNet50_Weights  # Specific model import

# Step 2: Mount Google Drive and Load Dataset
from google.colab import drive  # To access Google Drive

drive.mount('/content/gdrive')  # Mount Google Drive at given path

!unzip -q "/content/gdrive/MyDrive/ColabNotebooks_PegahKhosravi/CS74020_Spring2025/ucf101_top5.zip" -d /content/  # Unzip dataset

# Step 3: Set Parameters
IMG_SIZE = 112  # Resize frames to 112x112 pixels
BATCH_SIZE = 8  # Number of videos per batch
EPOCHS = 10  # Number of training epochs
MAX_SEQ_LENGTH = 20  # Max number of frames per video
NUM_FEATURES = 2048  # Output feature size from ResNet

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"  # Use GPU if available

# Step 4: Read Metadata
train_df = pd.read_csv("/content/ucf101_top5/train.csv")  # Load training data info
test_df = pd.read_csv("/content/ucf101_top5/test.csv")  # Load test data info

print(f"Training videos: {len(train_df)}")  # Print number of training samples
print(f"Testing videos: {len(test_df)}")  # Print number of testing samples

# Step 5: Define Video Loading Utilities
transform = transforms.Compose([
    transforms.ToPILImage(),  # Convert frame to PIL format
    transforms.Resize((IMG_SIZE, IMG_SIZE)),  # Resize frame
    transforms.ToTensor(),  # Convert to PyTorch tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # Normalize
])

def load_video(path, max_frames=MAX_SEQ_LENGTH):
    cap = cv2.VideoCapture(path)  # Open video file
    frames = []
    while True:
        ret, frame = cap.read()  # Read next frame
        if not ret:
            break  # Stop if no more frames
        frame = center_crop(frame)  # Crop to square
        frame = transform(frame)  # Apply preprocessing
        frames.append(frame)  # Store frame
        if len(frames) == max_frames:
            break  # Stop if max frames reached
    cap.release()  # Release video
    frames = torch.stack(frames)  # Stack into a tensor
    return frames  # Return tensor of shape (T, C, H, W)

def center_crop(frame):
    y, x, _ = frame.shape  # Get height and width
    min_dim = min(y, x)  # Use smaller dimension
    start_x = (x - min_dim) // 2  # Compute horizontal crop start
    start_y = (y - min_dim) // 2  # Compute vertical crop start
    return frame[start_y:start_y+min_dim, start_x:start_x+min_dim]  # Return cropped frame

# Step 6: Build Dataset Class
class VideoDataset(Dataset):  # Custom dataset class for video loading
    def __init__(self, df, root_dir):
        self.df = df  # Store dataframe with video paths and labels
        self.root_dir = root_dir  # Root folder with videos
        self.label2idx = {label: idx for idx, label in enumerate(sorted(df['tag'].unique()))}  # Map labels to integers

    def __len__(self):
        return len(self.df)  # Return number of items

    def __getitem__(self, idx):
        video_path = os.path.join(self.root_dir, self.df.iloc[idx]['video_name'])  # Full path to video
        label = self.label2idx[self.df.iloc[idx]['tag']]  # Get label index
        frames = load_video(video_path)  # Load preprocessed frames
        return frames, label  # Return input and label

# Step 7: Load Datasets
train_dataset = VideoDataset(train_df, "/content/ucf101_top5/train")  # Training set
test_dataset = VideoDataset(test_df, "/content/ucf101_top5/test")  # Test set

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2)  # Train data loader
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)  # Test data loader

# Step 8: Build Feature Extractor
class FeatureExtractor(nn.Module):  # Extract CNN features from frames
    def __init__(self):
        super().__init__()
        backbone = resnet50(weights=ResNet50_Weights.DEFAULT)  # Load pretrained ResNet50
        self.backbone = nn.Sequential(*list(backbone.children())[:-1])  # Remove final classification layer

    def forward(self, x):
        with torch.no_grad():  # Don't compute gradients
            features = self.backbone(x)  # Get features
            features = features.view(features.size(0), -1)  # Flatten features
        return features  # Return feature vector

feature_extractor = FeatureExtractor().to(DEVICE)  # Move to GPU/CPU
feature_extractor.eval()  # Set to evaluation mode

# Step 9: Build Sequence Model
class SequenceModel(nn.Module):  # GRU-based model for video classification
    def __init__(self, num_classes):
        super().__init__()
        self.gru = nn.GRU(NUM_FEATURES, 256, num_layers=2, batch_first=True)  # 2-layer GRU
        self.dropout = nn.Dropout(0.5)  # Dropout for regularization
        self.fc = nn.Linear(256, num_classes)  # Final classifier

    def forward(self, x):
        out, _ = self.gru(x)  # Run GRU
        out = out[:, -1, :]  # Get output at final timestep
        out = self.dropout(out)  # Apply dropout
        out = self.fc(out)  # Final classification
        return out

num_classes = len(train_df['tag'].unique())  # Count of unique labels
model = SequenceModel(num_classes).to(DEVICE)  # Instantiate model and move to device

# Step 10: Define Loss and Optimizer
criterion = nn.CrossEntropyLoss()  # Loss function
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)  # Optimizer

# Step 11: Training Loop
for epoch in range(EPOCHS):
    model.train()  # Set model to training mode
    running_loss = 0
    correct = 0
    total = 0

    for frames, labels in train_loader:
        batch_size, T, C, H, W = frames.size()  # Get shape
        frames = frames.view(-1, C, H, W).to(DEVICE)  # Reshape to process all frames at once
        labels = labels.to(DEVICE)

        features = feature_extractor(frames)  # Extract CNN features
        features = features.view(batch_size, T, -1)  # Reshape for GRU input

        outputs = model(features)  # Forward pass

        loss = criterion(outputs, labels)  # Compute loss

        optimizer.zero_grad()  # Zero gradients
        loss.backward()  # Backpropagation
        optimizer.step()  # Update weights

        running_loss += loss.item()  # Track loss
        _, preds = torch.max(outputs, 1)  # Get predicted classes
        correct += (preds == labels).sum().item()  # Count correct predictions
        total += labels.size(0)  # Total samples

    print(f"Epoch {epoch+1}/{EPOCHS}, Loss: {running_loss/len(train_loader):.4f}, Accuracy: {correct/total*100:.2f}%")

# Step 12: Evaluation
model.eval()  # Set model to eval mode
correct = 0
total = 0

with torch.no_grad():  # Disable gradient tracking
    for frames, labels in test_loader:
        batch_size, T, C, H, W = frames.size()
        frames = frames.view(-1, C, H, W).to(DEVICE)
        labels = labels.to(DEVICE)

        features = feature_extractor(frames)
        features = features.view(batch_size, T, -1)

        outputs = model(features)
        _, preds = torch.max(outputs, 1)

        correct += (preds == labels).sum().item()
        total += labels.size(0)

print(f"Test Accuracy: {correct/total*100:.2f}%")  # Print final test accuracy

# ---------------------------------------------------------------
# End of Code
# ---------------------------------------------------------------
````

This section visualizes the model's prediction on a single test video, along with a natural-color animation of its frames. First, a random test video is selected, loaded, and preprocessed into a sequence of normalized frames. These frames are passed through the pre-trained ResNet50 feature extractor and then through the trained GRU-based sequence model to predict the action label. The predicted class is mapped back to its original label name and printed. For visualization, the video frames are denormalized (to restore natural colors), converted to a NumPy format, and saved as a short GIF. This GIF is then displayed directly in the notebook to give users an intuitive view of the action being classified. This step helps validate the model’s predictions qualitatively by aligning them with what the user sees visually.

````python
# Prediction Visualization for a Single Test Video with Natural Colors
# --------------------------------------------------

import random  # To randomly select a video
import imageio  # For saving video frames as GIF
from IPython.display import Image, display  # For displaying images/GIFs in notebooks

# Step 13: Utility Functions to Visualize Prediction

def prepare_single_video(path):
    frames = load_video(path)  # Load and preprocess frames from the given video path
    frames = frames.unsqueeze(0)  # Add batch dimension; shape becomes (1, T, C, H, W)
    return frames  # Return a single-video batch

def predict_single_video(video_path):
    model.eval()  # Set model to evaluation mode
    feature_extractor.eval()  # Set feature extractor to evaluation mode

    frames = prepare_single_video(video_path)  # Load and batch the video
    batch_size, T, C, H, W = frames.size()  # Get shape
    frames = frames.view(-1, C, H, W).to(DEVICE)  # Reshape to feed into CNN

    with torch.no_grad():  # Disable gradient calculation
        features = feature_extractor(frames)  # Extract CNN features
        features = features.view(batch_size, T, -1)  # Reshape for GRU input
        outputs = model(features)  # Run through GRU-based classifier
        probs = torch.softmax(outputs, dim=1)  # Convert outputs to probabilities
        preds = torch.argmax(probs, dim=1)  # Get predicted class index

    idx_to_label = {v: k for k, v in train_dataset.label2idx.items()}  # Reverse label mapping
    predicted_label = idx_to_label[preds.item()]  # Convert index to label

    print(f"Predicted Action: {predicted_label}")  # Print the prediction

    return predicted_label  # Return label string

# Utility to denormalize frames for proper visualization

def denormalize(tensor):
    mean = torch.tensor([0.485, 0.456, 0.406]).view(1, 1, 3)  # Mean used during normalization
    std = torch.tensor([0.229, 0.224, 0.225]).view(1, 1, 3)  # Std used during normalization
    tensor = tensor * std + mean  # Invert normalization
    return tensor  # Return pixel values in [0, 1]

# Step 14: Show Test Video + Prediction

def show_video(frames_tensor):
    frames = frames_tensor.squeeze(0).permute(0, 2, 3, 1).cpu()  # Rearrange to (T, H, W, C)

    frames = denormalize(frames)  # Undo normalization for display
    frames = torch.clamp(frames, 0, 1)  # Clamp pixel values to valid range [0, 1]

    frames = (frames.numpy() * 255).astype(np.uint8)  # Convert to uint8 image array

    gif_path = "/content/temp_video.gif"  # Output path for GIF
    imageio.mimsave(gif_path, frames, fps=5)  # Save frames as animated GIF
    display(Image(filename=gif_path))  # Display the GIF in notebook

# Choose random test video
sample_video = random.choice(test_df['video_name'].values)  # Pick a random video name from test set
sample_video_path = os.path.join("/content/ucf101_top5/test", sample_video)  # Build full path

print(f"Sample Video Path: {sample_video}")  # Print selected video path
predict_single_video(sample_video_path)  # Predict the action in the selected video

# Show the video
frames_for_display = prepare_single_video(sample_video_path)  # Prepare frames for visualization
show_video(frames_for_display)  # Show the video as a GIF

# --------------------------------------------------
# End of Prediction Visualization
# --------------------------------------------------
````

## 1. Introduction to NLP

### What Makes NLP Hard?
Natural language is incredibly complex, ambiguous, and context-dependent. Some challenges include:
- **Ambiguity:** Words can have multiple meanings (e.g., "bank" as in riverbank vs. financial bank).
- **Context:** Meaning often depends on surrounding words or prior knowledge.
- **Variability:** There are many ways to express the same idea.
- **Structure:** Language follows grammar rules, but humans often bend them.
- **Data Sparsity:** Some word combinations are rare or unseen in training data. For example, even if you have a huge corpus, you might never see a phrase like "quantum banana synchronization", so your model has no direct experience with it. It's sparse in the sense that the space of all possible sentences or word combinations is huge, and the training data only covers a tiny fraction.

### Tokenization, Sequences, Vocabulary, and Context
Before feeding text into a machine learning model, we need to convert it into a numerical format that the model can understand. This involves several key steps:

- Tokenization: This is the process of breaking down raw text into smaller units called tokens, which could be words, subwords, or characters.

  - Example: "I love NLP!" → ["I", "love", "NLP", "!"]

- Sequences: Tokens are arranged in the same order they appear in the text, forming sequences. Sequence order is critical in NLP because it captures the structure and meaning of the sentence—unlike traditional ML where features are often unordered.

- Vocabulary: This is the collection of all unique tokens that the model is allowed to recognize. Each token is mapped to a unique integer ID.

  - Example: "love" → 17
  
    The actual IDs depend on how the vocabulary is constructed from the training data. Unless explicitly fixed, these mappings can vary between training runs.

- Context: Words derive meaning from surrounding words. For example, the word "apple" can refer to a fruit or a tech company, depending on its context. Modern NLP models are designed to use this context to better understand word meanings.

These core concepts form the basis for working with text in NLP and are crucial for understanding how more advanced models—like RNNs, attention mechanisms, and transformers—process language.



In this code, we demonstrate how to tokenize text using a pre-trained BERT tokenizer from Hugging Face’s Transformers library. We begin by importing the AutoTokenizer class and loading the tokenizer associated with the bert-base-uncased model, which is a lowercase, English version of BERT. We then define a sample input sentence, "I love bananas.", and use the tokenizer to break it down into smaller units called tokens (typically words or subwords). Finally, these tokens are mapped to their corresponding integer IDs, which represent their positions in BERT’s vocabulary. These integer IDs are what will actually be fed into the BERT model for further processing.

````python
# Import the AutoTokenizer class from Hugging Face Transformers
from transformers import AutoTokenizer

# Load the pretrained BERT tokenizer (public model, no login needed)
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Define the input text
text = "I love bananas."

# Tokenize the input text into tokens (words or subwords)
tokens = tokenizer.tokenize(text)
print("Tokens:", tokens)

# Convert the tokens into their corresponding integer IDs
input_ids = tokenizer.convert_tokens_to_ids(tokens)
print("Input IDs:", input_ids)
````

This output shows how the BERT tokenizer processes and encodes text. The sentence "I love bananas." is first split into tokens: ['i', 'love', 'bananas', '.']. These are lowercase word-level or subword-level tokens, as expected from the bert-base-uncased model. Each token is then mapped to its corresponding integer ID based on BERT’s vocabulary. For example, 'i' is mapped to 1045, 'love' to 2293, 'bananas' to 26191, and '.' to 1012. These input IDs are what BERT actually uses as input to perform tasks like classification, question answering, or text generation.

## 2. Word Embeddings
In Natural Language Processing (NLP), we need to convert words into numerical form so that machine learning models can process them. One of the most effective and meaningful ways to represent words numerically is through word embeddings.

### 2.1 From One-Hot Encoding to Word Embeddings
- One-Hot Encoding: In traditional representations, each word is encoded as a long vector of zeros with a single one in the position unique to that word.

  - Example: "cat" → [1, 0, 0, 0, 0, ...]
  - Limitation: These vectors are sparse, high-dimensional, and don't capture meaning — for the model, “cat” and “dog” look no more similar than “cat” and “spoon”.

- Word Embeddings (Distributed Representations):
Instead of one-hot vectors, embeddings represent each word as a dense vector in a lower-dimensional space (e.g., 100 or 300 dimensions). These vectors are learned such that words with similar meanings are close together in that space.

  - Example: Vectors for "king" and "queen" would be near each other and show relationships like gender or royalty.

### 2.2 Trainable Random Embeddings
When training a model from scratch, embeddings can start as random vectors and be learned during training. This is typically done using layers like nn.Embedding in PyTorch. Over time, these embeddings are updated using backpropagation to better reflect the task the model is learning (e.g., sentiment analysis or text classification).

### 2.3 Word2Vec: Learning Meaning from Context
Word2Vec is a popular algorithm that learns embeddings using a shallow neural network trained on massive text data. It comes in two main flavors:

- Skip-Gram: Predicts surrounding (context) words from a center word.

- CBOW (Continuous Bag of Words): Predicts the center word from its surrounding words.

These approaches capture local context and allow embeddings to encode semantic and syntactic relationships between words.

### 2.4 GloVe: Embeddings from Co-occurrence
GloVe (Global Vectors for Word Representation) takes a different approach. Instead of predicting words from context like Word2Vec, it builds a large matrix of word co-occurrence statistics (i.e., how often words appear together) and factorizes it to generate embeddings.

- This allows GloVe to combine global information (entire corpus statistics) with local context for better performance on certain tasks.

### 2.5 Using Pretrained Embeddings

Instead of learning embeddings from scratch, we can leverage embeddings that have already been trained on massive datasets like Wikipedia or Google News. Common options include:

- GloVe
- Word2Vec
- fastText

Benefits of pretrained embeddings:

- Faster training, especially with limited data
- Improved performance due to richer word representations
- Useful in transfer learning or as a foundation for more complex models

These embeddings are foundational to many modern NLP models — including RNNs, GRUs, LSTMs, and transformers like BERT — which build upon them to model entire sentences, paragraphs, and documents.

## 3. Recurrent Neural Networks (RNNs)

Recurrent Neural Networks (RNNs) are a foundational architecture for processing sequential data, such as text, speech, or time series. While newer models like Transformers now dominate NLP, understanding RNNs remains important — especially because they are still used in practical applications where memory efficiency or real-time inference matters.

### 3.1 Why Sequence Models Matter in NLP (and Beyond)
Many NLP tasks depend on context and word order. For example, the meaning of the word “bank” in the sentence “He went to the bank” depends on the surrounding words. Traditional neural networks treat each input independently and can’t remember what came before. RNNs were designed to address this by introducing a hidden state that carries information across time steps.

### 3.2 How RNNs Work (Quick Recap)

At each time step, the RNN takes the current input token and the previous hidden state to compute a new hidden state:

$$ h_t = \tanh(W_{hh} h_{t-1} + W_{xh} x_t + b) $$

This allows the model to “remember” previous inputs when generating predictions. However, basic RNNs struggle with long-term dependencies due to issues like vanishing gradients. This led to the development of more robust architectures such as LSTM and GRU, which handle longer sequences more effectively.

### 3.3 RNNs Beyond Text: Sequence Modeling in Video
While RNNs originated in NLP, their ability to model sequential data makes them useful in other domains, too. In our video classification project, for example, we use a CNN to extract spatial features from each video frame and a GRU (a type of RNN) to model the temporal dynamics — that is, how the action unfolds over time. This kind of sequence modeling is a direct extension of how RNNs were first used in NLP, but applied to video frames instead of words.

This code demonstrates how a simple Recurrent Neural Network (RNN) processes a sequence of characters using PyTorch. The input is the word "hello", which is first converted into one-hot encoded vectors for each character based on a small vocabulary of unique characters. A basic RNN layer is defined with an input size equal to the vocabulary size and a hidden state size of 8. The RNN takes the sequence of one-hot vectors as input and processes them one character at a time, updating its internal hidden state at each time step. The code prints the final hidden state after the full sequence is processed, as well as the intermediate hidden states at each step. This helps illustrate how an RNN maintains and evolves memory across a sequence, a fundamental concept behind sequence modeling in NLP and other domains.

````python
import torch
import torch.nn as nn

# Sample data: A toy sentence split into characters
sentence = "hello"
char_to_idx = {ch: idx for idx, ch in enumerate(sorted(set(sentence)))}
idx_to_char = {idx: ch for ch, idx in char_to_idx.items()}

# Hyperparameters
input_size = len(char_to_idx)  # input dimension (one-hot size)
hidden_size = 8  # RNN hidden state size
seq_length = len(sentence)  # Length of the sequence

# One-hot encoding for input
input_data = torch.zeros(seq_length, 1, input_size)
for t, ch in enumerate(sentence):
    input_data[t][0][char_to_idx[ch]] = 1

# Define a simple RNN
rnn = nn.RNN(input_size=input_size, hidden_size=hidden_size, batch_first=False)
hidden_state = torch.zeros(1, 1, hidden_size)  # Initial hidden state

# Forward pass
output, hn = rnn(input_data, hidden_state)

print("Input characters:", list(sentence))
print("Final hidden state:\n", hn)

# Show the hidden state at each time step
for i, o in enumerate(output):
    print(f"Hidden state at time step {i} for '{sentence[i]}': {o.data.numpy()}")
````

## 4. Sequence-to-Sequence (Seq2Seq) Models

Sequence-to-Sequence (Seq2Seq) models are designed to handle tasks where both the input and output are sequences — and possibly of different lengths. They are a natural extension of RNNs and are widely used in applications like machine translation, summarization, and dialogue systems.

### The Encoder-Decoder Framework

At the core of Seq2Seq models is the encoder-decoder architecture:

- **Encoder**: Reads the entire input sequence (e.g., a sentence in English) and compresses it into a fixed-length context vector — a numerical summary that captures the input’s meaning.
- **Decoder**: Takes this context vector and generates the output sequence (e.g., the translation in Persian), one token at a time.

The model is trained end-to-end to maximize the likelihood of producing the correct output sequence given the input.

### Real-World Applications

Seq2Seq models are used in many NLP and AI tasks, including:
- Machine Translation (e.g., English → Persian)
- Text Summarization
- Speech Recognition
- Chatbots and Dialogue Generation

### Teacher Forcing vs. Inference

**Teacher Forcing** is a training strategy where the decoder is given the ground-truth token from the previous time step as input — not its own prediction.  
This helps the model learn faster and more accurately, especially early in training.

> Example during training:  
> `Input: "I am"` → Target: `"من هستم"`  
> At time step 2, the decoder is fed `"من"` (correct word), even if it predicted something else at step 1.

**Inference**, on the other hand, reflects the real-world scenario: the model must rely on its own previous predictions to generate the next word.  
This makes generation harder and introduces the risk of compounding errors — one mistake early on can lead to a chain of incorrect outputs.

> Example during inference:  
> If the model predicts `"من"` at step 1, it feeds `"من"` into the next step, even if the correct word should’ve been something else.

This discrepancy between training and inference is a major challenge in sequence generation tasks and is actively addressed by techniques like scheduled sampling, beam search, and attention mechanisms.

### 5. Attention Mechanism

The attention mechanism was introduced to overcome a key limitation of the basic Seq2Seq architecture — the reliance on a fixed-length context vector, which can become a bottleneck, especially for long input sequences.

#### Why Do We Need Attention?

- In traditional Seq2Seq models, the encoder compresses the entire input sequence into a single vector, which the decoder must rely on to generate every word in the output.  
- For long or complex sequences, this fixed-length vector often fails to capture all the necessary information.
- Attention solves this by allowing the decoder to dynamically “look back” at all the encoder’s hidden states when generating each output token.

This leads to better performance and more flexible models, especially in tasks like machine translation, text summarization, and question answering.

#### How Attention Works (Conceptually)

At every step of decoding, the model:
1. Computes a relevance score between the current decoder state (query) and each encoder hidden state (keys).
2. Applies a softmax function to convert these scores into attention weights.
3. Computes a context vector as a weighted sum of encoder hidden states (values), guided by these attention weights.
4. Combines the context vector with the decoder's current state to predict the next output token.

#### Simplified Attention Equation

$$
\text{Attention}(q, K, V) = \text{softmax}\left(\frac{qK^T}{\sqrt{d_k}}\right)V
$$

Where:
- \( q \): query (decoder hidden state)  
- \( K \): keys (encoder hidden states)  
- \( V \): values (encoder hidden states)  
- $\( d_k \$): dimension of the key vectors (used for scaling)

#### Visualization & Interpretability

- Attention can be visualized as a heatmap that shows which parts of the input the model is focusing on while generating each output word.
- These visualizations provide insight into model behavior, making attention-based models more interpretable than traditional black-box systems.

#### From RNNs to Transformers

The idea of attention became so powerful that it eventually replaced recurrence altogether. In Transformer models, self-attention mechanisms enable the model to process sequences in parallel and focus on relationships between any pair of words — forming the foundation of modern NLP architectures like BERT, GPT, and T5.

# The Evolution of Attention: From Bahdanau to Vaswani

Before 2014, machine translation models used encoder-decoder RNNs. These models compressed the entire input sentence into one fixed-size context vector.

- This method worked for short sentences.
- However, for longer sentences, performance dropped because a single vector could not capture all the necessary information.

**Dzmitry Bahdanau**, **Kyunghyun Cho**, and **Yoshua Bengio** introduced a critical improvement in their 2014 paper:

> Why should the decoder memorize everything in one vector? Why not allow it to focus dynamically on different parts of the input while generating each word?

This insight led to the development of the attention mechanism, where the decoder "looks back" at different parts of the input sequence at each step, much like a student glancing back at various sections of their notes during an exam.

Their work fundamentally changed how models handled longer sequences and greatly improved translation quality.


## The Transformer Revolution

In 2017, researchers at **Google Brain**, led by **Ashish Vaswani**, proposed an even bolder idea:

> If attention works so well with RNNs, what if we build a model that uses only attention, and no RNNs at all?

This led to the landmark paper "Attention Is All You Need".

They introduced the Transformer architecture, which uses self-attention layers exclusively, eliminating the need for recurrence entirely.

Initially met with skepticism, Transformers proved to be:
- Faster to train
- More scalable to larger datasets
- Better in quality for many tasks

Today, Transformers are the foundation of nearly all state-of-the-art AI models, including **BERT**, **GPT-2**, **GPT-3**, **GPT-4**, **T5**, and more.

> Bahdanau sparked the revolution with attention, and Vaswani launched it into orbit with the Transformer.

````python
# Seq2Seq with Attention and Training Loop (Fixed)

import torch  # Core PyTorch library
import torch.nn as nn  # Neural network components
from torch.utils.data import DataLoader  # Efficient batch data loading
import torch.nn.functional as F  # Functional API for neural nets
import random  # For randomness in teacher forcing and sampling

# ----------------------------
# Data Setup (Manual Tokenizer + Vocab)
# ----------------------------

# English–Persian sentence pairs for translation
pairs = [
    ("hello", "سلام"),
    ("how are you", "چطوری"),
    ("i am fine", "من خوبم"),
    ("thank you", "ممنون"),
    ("good morning", "صبح بخیر"),
    ("good night", "شب بخیر"),
    ("see you soon", "به زودی می‌بینمت"),
    ("i love you", "من دوستت دارم"),
    ("what is your name", "اسم شما چیست"),
    ("my name is john", "اسم من جان است"),
    ("where are you", "کجا هستی"),
    ("i am at home", "من در خانه هستم"),
    ("do you speak english", "آیا انگلیسی صحبت می‌کنی"),
    ("i do not understand", "من متوجه نمی‌شوم"),
    ("can you help me", "میتونی کمکم کنی"),
    ("i am lost", "گم شده‌ام"),
    ("how much is this", "این چند است"),
    ("i like music", "من موسیقی دوست دارم"),
    ("i like football", "من فوتبال دوست دارم"),
    ("i am hungry", "من گرسنه هستم"),
    ("i am thirsty", "من تشنه هستم"),
    ("where is the station", "ایستگاه کجاست"),
    ("i need a doctor", "من به دکتر نیاز دارم"),
    ("i am tired", "من خسته هستم"),
    ("have a nice day", "روز خوبی داشته باشید"),
    ("good luck", "موفق باشی"),
    ("see you tomorrow", "فردا می‌بینمت"),
    ("happy birthday", "تولدت مبارک"),
    ("congratulations", "تبریک می‌گویم"),
    ("welcome", "خوش آمدید"),
    ("sorry", "ببخشید"),
    ("you are my sunshine", "تو آفتاب زندگی منی"),
    ("the stars shine for you", "ستارگان برای تو می‌درخشند"),
    ("i miss you", "دلم برایت تنگ شده"),
    ("you are always in my heart", "تو همیشه در قلب منی"),
    ("may your dreams come true", "امیدوارم رویاهایت به حقیقت بپیوندند"),
    ("life is beautiful", "زندگی زیباست"),
    ("i believe in you", "به تو ایمان دارم"),
    ("the world is brighter with you", "دنیا با حضور تو روشن‌تر است"),
    ("never give up", "هرگز تسلیم نشو"),
    ("follow your heart", "دنبال دلت برو"),
    ("may your journey be filled with light", "سفرت پر از نور باد"),
    ("kindness is a universal language", "مهربانی زبانی جهانی است"),
    ("your smile lights up my world", "لبخندت دنیایم را روشن می‌کند"),
    ("drink wine, this life is eternal no more", "می نوش که عمر جاودانی نیست"),
    ("be happy now, tomorrow is unknown", "شاد باش که فردا نامعلوم است"),
    ("the world is a moment, a dream, a shadow", "دنیا لحظه‌ای، خوابی و سایه‌ای بیش نیست"),
    ("seize the moment, life passes swiftly", "دم را دریاب که زندگی به شتاب می‌گذرد"),
    ("we come like the wind, we go like the wind", "چون باد آمدیم و چون باد رفتیم"),
    ("life is a fleeting dream", "زندگی رویایی گذراست"),
    ("open your heart to joy", "دلت را به شادی بسپار"),
    ("each breath is a treasure", "هر نفس گنجی است"),
    ("plant seeds of kindness in the garden of time", "در باغ زمان بذر مهربانی بکار"),
    ("oh heart, seek joy before it fades", "ای دل، شادی بجوی پیش از آنکه بگذرد")

]

# Tokenizer that lowercases and splits a sentence into words
def tokenize(text):
    return text.lower().strip().split()

# Build vocabulary from a list of sentences
def build_vocab(sentences):
    vocab = {"<pad>": 0, "<sos>": 1, "<eos>": 2}  # Special tokens
    idx = 3
    for sentence in sentences:
        for token in tokenize(sentence):
            if token not in vocab:
                vocab[token] = idx
                idx += 1
    return vocab

# Split into source (English) and target (Persian) sentences
src_sentences = [src for src, _ in pairs]
trg_sentences = [trg for _, trg in pairs]

# Build vocabularies and inverse mappings
SRC_vocab = build_vocab(src_sentences)
TRG_vocab = build_vocab(trg_sentences)
SRC_itos = {i: s for s, i in SRC_vocab.items()}
TRG_itos = {i: s for s, i in TRG_vocab.items()}

# Convert sentence pairs into padded tensors with <sos> and <eos>
def tensorize(pair):
    src = [SRC_vocab["<sos>"]] + [SRC_vocab[token] for token in tokenize(pair[0])] + [SRC_vocab["<eos>"]]
    trg = [TRG_vocab["<sos>"]] + [TRG_vocab[token] for token in tokenize(pair[1])] + [TRG_vocab["<eos>"]]
    return torch.tensor(src), torch.tensor(trg)

data = [tensorize(pair) for pair in pairs]  # Full dataset

# Custom collate function for DataLoader padding
def collate_fn(batch):
    src_batch, trg_batch = zip(*batch)
    src_batch = nn.utils.rnn.pad_sequence(src_batch)  # Pad source sentences
    trg_batch = nn.utils.rnn.pad_sequence(trg_batch)  # Pad target sentences
    return src_batch, trg_batch

loader = DataLoader(data, batch_size=2, shuffle=True, collate_fn=collate_fn)  # Mini-batch loader

# ----------------------------
# Model Components
# ----------------------------

# Encoder: Embeds and encodes input sequence
class Encoder(nn.Module):
    def __init__(self, input_dim, emb_dim, hid_dim):
        super().__init__()
        self.embedding = nn.Embedding(input_dim, emb_dim)
        self.rnn = nn.GRU(emb_dim, hid_dim)

    def forward(self, src):
        embedded = self.embedding(src)  # [src_len, batch_size, emb_dim]
        outputs, hidden = self.rnn(embedded)  # outputs: [src_len, batch_size, hid_dim]
        return outputs, hidden

# Attention mechanism: computes attention weights
class Attention(nn.Module):
    def __init__(self, hid_dim):
        super().__init__()
        self.attn = nn.Linear(hid_dim * 2, 1)  # Combine encoder output and hidden state

    def forward(self, hidden, encoder_outputs):
        src_len = encoder_outputs.shape[0]
        hidden = hidden.repeat(src_len, 1, 1)  # Repeat decoder hidden for each time step
        energy = torch.tanh(self.attn(torch.cat((hidden, encoder_outputs), dim=2)))  # [src_len, batch, 1]
        attention = torch.softmax(energy.squeeze(2), dim=0)  # Normalize over source length
        return attention  # [src_len, batch]

# Decoder with attention
class Decoder(nn.Module):
    def __init__(self, output_dim, emb_dim, hid_dim, attention):
        super().__init__()
        self.embedding = nn.Embedding(output_dim, emb_dim)
        self.rnn = nn.GRU(hid_dim + emb_dim, hid_dim)
        self.fc_out = nn.Linear(hid_dim * 2, output_dim)
        self.attention = attention

    def forward(self, input, hidden, encoder_outputs):
        input = input.unsqueeze(0)  # Make input shape [1, batch_size]
        embedded = self.embedding(input)  # [1, batch, emb_dim]

        # Compute attention weights
        a = self.attention(hidden, encoder_outputs)  # [src_len, batch]
        a = a.permute(1, 0).unsqueeze(1)  # [batch, 1, src_len]
        encoder_outputs = encoder_outputs.permute(1, 0, 2)  # [batch, src_len, hid_dim]
        weighted = torch.bmm(a, encoder_outputs)  # [batch, 1, hid_dim]
        weighted = weighted.permute(1, 0, 2)  # [1, batch, hid_dim]

        rnn_input = torch.cat((embedded, weighted), dim=2)  # [1, batch, emb+hid]
        output, hidden = self.rnn(rnn_input, hidden)  # output: [1, batch, hid_dim]
        prediction = self.fc_out(torch.cat((output.squeeze(0), weighted.squeeze(0)), dim=1))  # [batch, output_dim]

        return prediction, hidden

# ----------------------------
# Training Loop
# ----------------------------

# Full Seq2Seq model
class Seq2Seq(nn.Module):
    def __init__(self, encoder, decoder, device):
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.device = device

    def forward(self, src, trg, teacher_forcing_ratio=0.5):
        batch_size = trg.shape[1]
        trg_len = trg.shape[0]
        trg_vocab_size = self.decoder.fc_out.out_features

        outputs = torch.zeros(trg_len, batch_size, trg_vocab_size).to(self.device)  # Store predictions

        encoder_outputs, hidden = self.encoder(src)  # Encode source
        input = trg[0, :]  # First input to decoder is <sos>

        for t in range(1, trg_len):
            output, hidden = self.decoder(input, hidden, encoder_outputs)  # Decode step
            outputs[t] = output  # Store output
            top1 = output.argmax(1)  # Get highest-probability word
            # Decide whether to use the actual target word (teacher forcing) or the model's own prediction from the previous step
            # Generate a random number between 0 and 1; if it's less than the teacher forcing ratio, use the correct next word (trg[t]); otherwise, use the model's own prediction (top1)
            input = trg[t] if random.random() < teacher_forcing_ratio else top1  # Teacher forcing

        return outputs  # All decoder outputs

# ----------------------------
# Model Setup & Training
# ----------------------------

INPUT_DIM = len(SRC_vocab)  # Input vocabulary size
OUTPUT_DIM = len(TRG_vocab)  # Output vocabulary size
EMB_DIM = 64  # Embedding dimension
HID_DIM = 128  # Hidden state size
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")  # Use GPU if available

# Initialize components
enc = Encoder(INPUT_DIM, EMB_DIM, HID_DIM).to(DEVICE)
attn = Attention(HID_DIM).to(DEVICE)
dec = Decoder(OUTPUT_DIM, EMB_DIM, HID_DIM, attn).to(DEVICE)
model = Seq2Seq(enc, dec, DEVICE).to(DEVICE)

# Optimizer and loss function
optimizer = torch.optim.Adam(model.parameters())
criterion = nn.CrossEntropyLoss(ignore_index=TRG_vocab["<pad>"])  # Ignore padding in loss

EPOCHS = 100  # Training epochs
model.train()  # Set model to training mode

for epoch in range(EPOCHS):
    epoch_loss = 0  # Track loss

    for src_batch, trg_batch in loader:
        src_batch, trg_batch = src_batch.to(DEVICE), trg_batch.to(DEVICE)
        optimizer.zero_grad()  # Zero gradients
        output = model(src_batch, trg_batch)  # Forward pass

        output_dim = output.shape[-1]
        output = output[1:].reshape(-1, output_dim)  # Remove <sos> and flatten
        trg = trg_batch[1:].reshape(-1)  # Flatten target as well

        loss = criterion(output, trg)  # Compute loss
        loss.backward()  # Backpropagation
        optimizer.step()  # Update parameters

        epoch_loss += loss.item()  # Accumulate loss

    print(f"Epoch {epoch+1} | Loss: {epoch_loss/len(loader):.4f}")  # Log progress

print("Training complete!")  # Final confirmation
````

This code defines a function, translate_sentence(), that performs inference using the trained Seq2Seq model with attention. It takes an English input sentence, tokenizes it, converts the tokens into numerical indices using the source vocabulary, and feeds them into the encoder. Then, using the decoder and the attention mechanism, it generates one Persian token at a time, starting from the <sos> token. The decoding process continues until the model predicts an <eos> token or reaches the maximum allowed length. The predicted indices are converted back to Persian words and returned as a translated string. This allows the model to be used in real-world scenarios where it needs to translate sentences that were not explicitly part of the training set.

````python
# Translation Inference with Trained Seq2Seq Model (English ➔ Persian)

def translate_sentence(model, sentence, SRC_vocab, TRG_vocab, SRC_itos, TRG_itos, device, max_len=20):
    model.eval()  # Set the model to evaluation mode (no dropout, no gradient updates)

    # Tokenize the input sentence and add <sos> and <eos> tokens
    tokens = ["<sos>"] + tokenize(sentence) + ["<eos>"]

    # Convert tokens to indices using the source vocabulary
    src_indexes = [SRC_vocab.get(token, SRC_vocab["<pad>"]) for token in tokens]

    # Convert to tensor and add batch dimension; shape = [sequence_length, 1]
    src_tensor = torch.tensor(src_indexes).unsqueeze(1).to(device)

    # Pass source through encoder (no gradients needed)
    with torch.no_grad():
        encoder_outputs, hidden = model.encoder(src_tensor)

    trg_indexes = [TRG_vocab["<sos>"]]  # Start target sequence with <sos>

    # Iteratively generate tokens up to max_len
    for i in range(max_len):
        trg_tensor = torch.tensor([trg_indexes[-1]]).to(device)  # Get last predicted token
        with torch.no_grad():
            output, hidden = model.decoder(trg_tensor, hidden, encoder_outputs)  # Decode step

        pred_token = output.argmax(1).item()  # Get the index of the most probable next token
        trg_indexes.append(pred_token)  # Add predicted token to target sequence

        if pred_token == TRG_vocab["<eos>"]:  # Stop if <eos> is predicted
            break

    # Convert token indices to words, excluding special tokens
    trg_tokens = [TRG_itos[i] for i in trg_indexes if i not in [TRG_vocab["<sos>"], TRG_vocab["<eos>"], TRG_vocab["<pad>"]]]

    return " ".join(trg_tokens)  # Return the final translated sentence

# Example Usage
example_sentences = [  # English input sentences for translation
    "my name is Pegah",
    "life is a fleeting dream"
]

for sentence in example_sentences:
    # Translate each sentence and print the result
    translation = translate_sentence(model, sentence, SRC_vocab, TRG_vocab, SRC_itos, TRG_itos, DEVICE)
    print(f"EN: {sentence} → FA: {translation}")  # Show English ➔ Persian translation
````

When running the inference function on new input sentences, we observe that the model often outputs translations that closely match examples it has seen during training, rather than fully adapting to novel inputs. For instance, given "my name is Pegah", the model returns "اسم من جان است" ("My name is John") because it encountered that exact phrase in the training data and has memorized the pattern. Since “Pegah” is an unseen token, the model defaults to the closest familiar structure. This demonstrates a key limitation of training from scratch on a small dataset: the model tends to memorize rather than generalize, which restricts its ability to handle new vocabulary or structures beyond what it was explicitly trained on.

## Visualizing Attention Weights (Heatmap)

In this section, we visualize the attention mechanism at work inside the Seq2Seq model with manual Persian target alignment.
After generating a translation, we plot a heatmap showing how much attention the model paid to each source token when predicting each manually provided target token.

This code implements a custom function, visualize_attention_manual(), to visualize the attention weights of a trained Seq2Seq model during English-to-Persian translation. It tokenizes the input English sentence, encodes it, and then during the decoding phase, captures the attention scores at each target prediction step. The attention matrix, which reflects how much the model "focused" on each source token when generating each target token, is then plotted as a heatmap using matplotlib and seaborn. Special handling for Persian text is added through the arabic_reshaper and bidi libraries to ensure correct right-to-left display. The function takes in the manual Persian target sentence (to correctly align labels) and produces a visual map showing how the model attends to different parts of the input at each generation step — providing an intuitive way to understand and debug the model’s translation behavior.

````python
# Attention Visualization for English ➔ Persian

# Step 1: Install libraries (run this cell once)
!pip install arabic-reshaper
!pip install python-bidi
````

````python
# Attention Visualization

# Step 2: Import required libraries
import matplotlib.pyplot as plt  # For plotting
import seaborn as sns  # For nicer heatmaps
import numpy as np  # For array handling
import arabic_reshaper  # To reshape Persian text for correct display
from bidi.algorithm import get_display  # To display right-to-left Persian correctly
import torch  # PyTorch

# Step 3: Define utility function to reshape Persian tokens
def reshape_persian_token(token):
    reshaped_text = arabic_reshaper.reshape(token)  # Reshape text for proper Persian character joining
    bidi_text = get_display(reshaped_text)  # Display in correct right-to-left format
    return bidi_text  # Return displayable Persian token

# Simple tokenizer to split English and Persian sentences
def tokenize(text):
    return text.lower().strip().split()

# Step 4: Define the final attention visualization function
def visualize_attention_manual(model, src_sentence, tgt_sentence, SRC_vocab, TRG_vocab, SRC_itos, TRG_itos, device):
    model.eval()  # Set model to evaluation mode

    # Tokenize and index the source sentence (English)
    src_tokens = ["<sos>"] + tokenize(src_sentence) + ["<eos>"]
    src_indexes = [SRC_vocab.get(token, SRC_vocab.get("<pad>", 0)) for token in src_tokens]
    src_tensor = torch.tensor(src_indexes).unsqueeze(1).to(device)  # Shape: [seq_len, 1]

    with torch.no_grad():
        encoder_outputs, hidden = model.encoder(src_tensor)  # Encode source sequence

    trg_indexes = [TRG_vocab["<sos>"]]  # Start target sentence with <sos>
    attentions = []  # Store attention weights for each decoding step

    for _ in range(50):  # Generate up to 50 tokens in target (Persian)
        trg_tensor = torch.tensor([trg_indexes[-1]]).to(device)  # Get most recent token
        with torch.no_grad():
            output, hidden = model.decoder(trg_tensor, hidden, encoder_outputs)  # Decode step

        pred_token = output.argmax(1).item()  # Get predicted token
        trg_indexes.append(pred_token)  # Add to target sentence

        if pred_token == TRG_vocab["<eos>"]:  # Stop if <eos> is generated
            break

        # Compute attention weights for current step
        attn_weights = model.decoder.attention(hidden, encoder_outputs)
        attn_weights = attn_weights.squeeze(1).detach().cpu().numpy()  # Convert to NumPy
        attentions.append(attn_weights)  # Save attention vector

    if len(attentions) == 0:
        print("No attentions to plot!")  # Handle case with no attention recorded
        return

    attn_matrix = np.stack(attentions)  # Combine all attention vectors into matrix

    # Use the *provided* target sentence (Persian), not the model prediction
    tgt_tokens = tokenize(tgt_sentence)

    # Reshape Persian tokens for right-to-left display
    tgt_tokens_reshaped = [reshape_persian_token(token) for token in tgt_tokens]

    # Step 5: Plot the attention heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(
        attn_matrix[::-1],  # Flip matrix vertically so first target is on top
        xticklabels=src_tokens,  # English tokens
        yticklabels=tgt_tokens_reshaped[::-1],  # Persian tokens, also flipped
        cmap="coolwarm"  # Color style
    )

    plt.xlabel("Source (English)")
    plt.ylabel("Target (Persian)")
    plt.title("Attention Heatmap: English ➔ Persian (Manual Target Text)")
    plt.xticks(rotation=45)  # Rotate x-labels for readability
    plt.tight_layout()
    plt.show()

# Step 6: Try it out on a sample translation
visualize_attention_manual(
    model=model,
    src_sentence="i am at home",  # English input
    tgt_sentence="من در خانه هستم",  # Persian reference
    SRC_vocab=SRC_vocab,
    TRG_vocab=TRG_vocab,
    SRC_itos=SRC_itos,
    TRG_itos=TRG_itos,
    device=DEVICE
)
````

### What are `<sos>` and `<eos>` Tokens?

**Understanding <sos> and <eos> Tokens in Attention Maps**

In Seq2Seq models, two special tokens are added to help the model understand the structure of sentences:

- <sos> (Start of Sequence) — signals where the sentence begins.

- <eos> (End of Sequence) — signals where the sentence ends.

These tokens help the model know when to start and stop generating text during training and inference.

**Why Do <sos> and <eos> Appear in the Heatmap?**

In attention heatmaps:

- The X-axis shows the source sentence, including <sos> and <eos>, because they are part of what the encoder sees.

- The Y-axis shows the target words, excluding <sos>, since we usually only visualize the actual translation (not the internal decoding setup).

**Simple Analogy**

Think of <sos> as the green light at the start of a race and <eos> as the finish line — both are essential, but you don’t watch them throughout the race. 🏁


# Word Embeddings + Seq2Seq: English ➔ Persian Translation with GloVe

This code demonstrates how to combine **pre-trained GloVe word embeddings** with a sequence-to-sequence (Seq2Seq) translation model in PyTorch for translating English sentences into Persian.

It begins by downloading and parsing 100-dimensional GloVe vectors into a Python dictionary that maps words to their corresponding embedding vectors. After inspecting a few sample words and computing their cosine similarities to explore semantic relationships, the code constructs an embedding matrix for a small custom vocabulary. This matrix is then used to initialize a PyTorch nn.Embedding layer, allowing models to start with rich, meaningful word representations. Finally, a sample batch of word indices is passed through the embedding layer to produce word vectors, which can be fed into downstream models like RNNs or GRUs for sequence processing tasks.


````python
# Word Embeddings with GloVe: Loading, Exploring, and Using in PyTorch

# Step 1: Import necessary libraries
import torch
import torch.nn as nn
import numpy as np

# Step 2: Download and Extract GloVe Embeddings
# We use 100-dimensional GloVe vectors trained on 6B tokens from Wikipedia + Gigaword.
!wget -q http://nlp.stanford.edu/data/glove.6B.zip
!unzip -q glove.6B.zip -d glove

# Step 3: Load GloVe Vectors into Python
# We create a dictionary: {word: embedding_vector}

glove_path = "glove/glove.6B.100d.txt"
embeddings_index = {}

with open(glove_path, 'r', encoding='utf-8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], dtype='float32')
        embeddings_index[word] = vector

print(f"Loaded {len(embeddings_index):,} word vectors from GloVe.")

# Step 4: Explore Word Vectors and Their Similarities
# Define a function to compute cosine similarity between two vectors

def cosine_similarity(vec1, vec2):
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

# List some example words
words = ["king", "queen", "man", "woman"]
for w in words:
    print(f"{w}: vector shape = {embeddings_index[w].shape}")

# Compute and print cosine similarity between word pairs
print("\nCosine Similarities:")
pairs = [("king", "queen"), ("man", "woman"), ("king", "man"), ("queen", "woman")]
for w1, w2 in pairs:
    sim = cosine_similarity(embeddings_index[w1], embeddings_index[w2])
    print(f"{w1} ↔ {w2} = {sim:.4f}")

# Step 5: Build a PyTorch Embedding Layer Using GloVe
# Assume we have a small vocabulary from our dataset

vocab = {"<pad>": 0, "<sos>": 1, "<eos>": 2, "king": 3, "queen": 4, "man": 5, "woman": 6, "hello": 7, "you": 8}
embedding_dim = 100
vocab_size = len(vocab)

# Create an embedding matrix initialized from GloVe or randomly if OOV (Out-of-Vocabulary)
embedding_matrix = np.zeros((vocab_size, embedding_dim))
for word, idx in vocab.items():
    vector = embeddings_index.get(word)
    if vector is not None:
        embedding_matrix[idx] = vector
    else:
        embedding_matrix[idx] = np.random.normal(scale=0.6, size=(embedding_dim,))  # Random init fallback

# Build PyTorch Embedding Layer
embedding_layer = nn.Embedding(vocab_size, embedding_dim)
embedding_layer.weight.data.copy_(torch.from_numpy(embedding_matrix))
embedding_layer.weight.requires_grad = True  # Optional: allow training or freeze

# Step 6: Example Usage: Embedding a batch of word indices
sample_batch = torch.tensor([[3, 4, 5], [6, 7, 8]])  # [batch_size=2, sequence_length=3]
embedded = embedding_layer(sample_batch)

print(f"\nEmbedded shape: {embedded.shape}")  # Should be [2, 3, 100]
````

Cosine similarity scores demonstrate meaningful semantic relationships: "man" and "woman" have the highest similarity (0.8323), followed by "king" and "queen" (0.7508), indicating that GloVe captures gender and role-based associations. Finally, when a sample batch of word indices is passed through the embedding layer, the resulting tensor has the shape [2, 3, 100], confirming that each word in the batch is represented by a 100-dimensional vector as expected.

In this example, a simple GRU (Gated Recurrent Unit) layer is defined and used to process the embedded word vectors. The GRU takes input with a shape of [batch_size, sequence_length, embedding_dim], which matches the shape of the output from the embedding layer. When the embedded sample batch is passed through the GRU, it returns two outputs: output and hidden. The output tensor contains the hidden states for each time step in the sequence, while the hidden tensor captures the final hidden state for each sequence in the batch. The printed shapes confirm this: output has shape [2, 3, 64], representing two sequences of three tokens each, encoded into 64-dimensional vectors; hidden has shape [1, 2, 64], indicating one layer, two sequences, and 64 hidden units per sequence.

````python
# Example: Pass embeddings into a simple GRU

# Define a simple GRU layer
gru_layer = nn.GRU(input_size=embedding_dim, hidden_size=64, batch_first=True)

# Pass the embedded sample batch through the GRU
output, hidden = gru_layer(embedded)

print(f"GRU Output Shape: {output.shape}")  # [batch_size, sequence_len, hidden_size]
print(f"GRU Hidden State Shape: {hidden.shape}")  # [num_layers, batch_size, hidden_size]
````

When the embedded batch of word vectors is passed through the GRU layer, the model returns two outputs. The first, output, has a shape of [2, 3, 64], which corresponds to a batch size of 2, sequence length of 3, and a GRU hidden size of 64. This means the GRU produces a 64-dimensional hidden state for each token in each sequence. The second output, hidden, has a shape of [1, 2, 64], indicating that it contains the final hidden state for each sequence in the batch (2 sequences), with 1 GRU layer and 64 hidden units. This result confirms that the GRU correctly processed the input embeddings and is ready to be integrated into a full sequence model such as an encoder in a Seq2Seq architecture.

````python
# English ➔ Persian Translation with GloVe Embeddings (Clean Version)

# Step 1: Import Libraries
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
import numpy as np
import random

# Step 2: # 📥 Load GloVe Embeddings (Only Run Once per Session)

# Step 3: Dataset (English to Persian)
pairs = [
    ("hello", "سلام"),
    ("how are you", "چطوری"),
    ("i am fine", "من خوبم"),
    ("thank you", "ممنون"),
    ("good morning", "صبح بخیر"),
    ("good night", "شب بخیر"),
    ("see you soon", "به زودی می‌بینمت"),
    ("i love you", "من دوستت دارم"),
    ("what is your name", "اسم شما چیست"),
    ("my name is john", "اسم من جان است"),
    ("where are you", "کجا هستی"),
    ("i am at home", "من در خانه هستم"),
    ("do you speak english", "آیا انگلیسی صحبت می‌کنی"),
    ("i do not understand", "من متوجه نمی‌شوم"),
    ("can you help me", "میتونی کمکم کنی"),
    ("i am lost", "گم شده‌ام"),
    ("how much is this", "این چند است"),
    ("i like music", "من موسیقی دوست دارم"),
    ("i like football", "من فوتبال دوست دارم"),
    ("i am hungry", "من گرسنه هستم"),
    ("i am thirsty", "من تشنه هستم"),
    ("where is the station", "ایستگاه کجاست"),
    ("i need a doctor", "من به دکتر نیاز دارم"),
    ("i am tired", "من خسته هستم"),
    ("have a nice day", "روز خوبی داشته باشید"),
    ("good luck", "موفق باشی"),
    ("see you tomorrow", "فردا می‌بینمت"),
    ("happy birthday", "تولدت مبارک"),
    ("congratulations", "تبریک می‌گویم"),
    ("welcome", "خوش آمدید"),
    ("sorry", "ببخشید"),
    ("you are my sunshine", "تو آفتاب زندگی منی"),
    ("the stars shine for you", "ستارگان برای تو می‌درخشند"),
    ("i miss you", "دلم برایت تنگ شده"),
    ("you are always in my heart", "تو همیشه در قلب منی"),
    ("may your dreams come true", "امیدوارم رویاهایت به حقیقت بپیوندند"),
    ("life is beautiful", "زندگی زیباست"),
    ("i believe in you", "به تو ایمان دارم"),
    ("the world is brighter with you", "دنیا با حضور تو روشن‌تر است"),
    ("never give up", "هرگز تسلیم نشو"),
    ("follow your heart", "دنبال دلت برو"),
    ("may your journey be filled with light", "سفرت پر از نور باد"),
    ("kindness is a universal language", "مهربانی زبانی جهانی است"),
    ("your smile lights up my world", "لبخندت دنیایم را روشن می‌کند"),
    ("drink wine, this life is eternal no more", "می نوش که عمر جاودانی نیست"),
    ("be happy now, tomorrow is unknown", "شاد باش که فردا نامعلوم است"),
    ("the world is a moment, a dream, a shadow", "دنیا لحظه‌ای، خوابی و سایه‌ای بیش نیست"),
    ("seize the moment, life passes swiftly", "دم را دریاب که زندگی به شتاب می‌گذرد"),
    ("we come like the wind, we go like the wind", "چون باد آمدیم و چون باد رفتیم"),
    ("life is a fleeting dream", "زندگی رویایی گذراست"),
    ("open your heart to joy", "دلت را به شادی بسپار"),
    ("each breath is a treasure", "هر نفس گنجی است"),
    ("plant seeds of kindness in the garden of time", "در باغ زمان بذر مهربانی بکار"),
    ("oh heart, seek joy before it fades", "ای دل، شادی بجوی پیش از آنکه بگذرد")

]

def tokenize(text):
    return text.lower().strip().split()

def build_vocab(sentences):
    vocab = {"<pad>": 0, "<sos>": 1, "<eos>": 2}
    idx = 3
    for sentence in sentences:
        for token in tokenize(sentence):
            if token not in vocab:
                vocab[token] = idx
                idx += 1
    return vocab

src_sentences = [src for src, _ in pairs]
trg_sentences = [trg for _, trg in pairs]
SRC_vocab = build_vocab(src_sentences)
TRG_vocab = build_vocab(trg_sentences)
SRC_itos = {i: s for s, i in SRC_vocab.items()}
TRG_itos = {i: s for s, i in TRG_vocab.items()}

# Step 4: Build embedding matrix for English (source side)
vocab_size = len(SRC_vocab)
embedding_dim = 100
embedding_matrix = np.zeros((vocab_size, embedding_dim))
for word, idx in SRC_vocab.items():
    vector = embeddings_index.get(word)
    if vector is not None:
        embedding_matrix[idx] = vector
    else:
        embedding_matrix[idx] = np.random.normal(scale=0.6, size=(embedding_dim,))

# Step 5: Create Dataset

def tensorize(pair):
    src = [SRC_vocab["<sos>"]] + [SRC_vocab[token] for token in tokenize(pair[0])] + [SRC_vocab["<eos>"]]
    trg = [TRG_vocab["<sos>"]] + [TRG_vocab[token] for token in tokenize(pair[1])] + [TRG_vocab["<eos>"]]
    return torch.tensor(src), torch.tensor(trg)

data = [tensorize(pair) for pair in pairs]

def collate_fn(batch):
    src_batch, trg_batch = zip(*batch)
    src_batch = nn.utils.rnn.pad_sequence(src_batch)
    trg_batch = nn.utils.rnn.pad_sequence(trg_batch)
    return src_batch, trg_batch

loader = DataLoader(data, batch_size=2, shuffle=True, collate_fn=collate_fn)

# Step 6: Define the Model (Encoder, Attention, Decoder, Seq2Seq)
class Encoder(nn.Module):
    def __init__(self, vocab_size, emb_dim, hid_dim, embedding_weights):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, emb_dim)
        self.embedding.weight.data.copy_(torch.from_numpy(embedding_weights))
        self.embedding.weight.requires_grad = True
        self.rnn = nn.GRU(emb_dim, hid_dim)

    def forward(self, src):
        embedded = self.embedding(src)
        outputs, hidden = self.rnn(embedded)
        return outputs, hidden

class Attention(nn.Module):
    def __init__(self, hid_dim):
        super().__init__()
        self.attn = nn.Linear(hid_dim * 2, 1)

    def forward(self, hidden, encoder_outputs):
        src_len = encoder_outputs.shape[0]
        hidden = hidden.repeat(src_len, 1, 1)
        energy = torch.tanh(self.attn(torch.cat((hidden, encoder_outputs), dim=2)))
        attention = torch.softmax(energy.squeeze(2), dim=0)
        return attention

class Decoder(nn.Module):
    def __init__(self, output_dim, emb_dim, hid_dim, attention):
        super().__init__()
        self.embedding = nn.Embedding(output_dim, emb_dim)
        self.rnn = nn.GRU(hid_dim + emb_dim, hid_dim)
        self.fc_out = nn.Linear(hid_dim * 2, output_dim)
        self.attention = attention

    def forward(self, input, hidden, encoder_outputs):
        input = input.unsqueeze(0)
        embedded = self.embedding(input)
        a = self.attention(hidden, encoder_outputs)
        a = a.permute(1, 0).unsqueeze(1)
        encoder_outputs = encoder_outputs.permute(1, 0, 2)
        weighted = torch.bmm(a, encoder_outputs)
        weighted = weighted.permute(1, 0, 2)
        rnn_input = torch.cat((embedded, weighted), dim=2)
        output, hidden = self.rnn(rnn_input, hidden)
        prediction = self.fc_out(torch.cat((output.squeeze(0), weighted.squeeze(0)), dim=1))
        return prediction, hidden

class Seq2Seq(nn.Module):
    def __init__(self, encoder, decoder, device):
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.device = device

    def forward(self, src, trg, teacher_forcing_ratio=0.5):
        batch_size = trg.shape[1]
        trg_len = trg.shape[0]
        trg_vocab_size = self.decoder.fc_out.out_features

        outputs = torch.zeros(trg_len, batch_size, trg_vocab_size).to(self.device)
        encoder_outputs, hidden = self.encoder(src)
        input = trg[0, :]

        for t in range(1, trg_len):
            output, hidden = self.decoder(input, hidden, encoder_outputs)
            outputs[t] = output
            top1 = output.argmax(1)
            input = trg[t] if random.random() < teacher_forcing_ratio else top1

        return outputs

# Step 7: Instantiate Model
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
INPUT_DIM = len(SRC_vocab)
OUTPUT_DIM = len(TRG_vocab)
EMB_DIM = 100
HID_DIM = 128

enc = Encoder(INPUT_DIM, EMB_DIM, HID_DIM, embedding_matrix).to(DEVICE)
attn = Attention(HID_DIM).to(DEVICE)
dec = Decoder(OUTPUT_DIM, EMB_DIM, HID_DIM, attn).to(DEVICE)
model = Seq2Seq(enc, dec, DEVICE).to(DEVICE)

optimizer = torch.optim.Adam(model.parameters())
criterion = nn.CrossEntropyLoss(ignore_index=TRG_vocab["<pad>"])

# Step 8: Train the Model
EPOCHS = 100

model.train()
for epoch in range(EPOCHS):
    epoch_loss = 0
    for src_batch, trg_batch in loader:
        src_batch, trg_batch = src_batch.to(DEVICE), trg_batch.to(DEVICE)
        optimizer.zero_grad()
        output = model(src_batch, trg_batch)
        output_dim = output.shape[-1]
        output = output[1:].view(-1, output_dim)
        trg = trg_batch[1:].reshape(-1)

        loss = criterion(output, trg)
        loss.backward()
        optimizer.step()

        epoch_loss += loss.item()

    print(f"Epoch {epoch+1} | Loss: {epoch_loss/len(loader):.4f}")

print("Training complete with GloVe embeddings!")
````

````python
# Translation Inference after Training

def translate_sentence(model, sentence, SRC_vocab, TRG_vocab, SRC_itos, TRG_itos, device, max_len=20):
    model.eval()

    # Step 1: Tokenize and numericalize input
    tokens = ["<sos>"] + tokenize(sentence) + ["<eos>"]
    src_indexes = [SRC_vocab.get(token, SRC_vocab["<pad>"]) for token in tokens]
    src_tensor = torch.tensor(src_indexes).unsqueeze(1).to(device)

    # Step 2: Encode source sentence
    with torch.no_grad():
        encoder_outputs, hidden = model.encoder(src_tensor)

    # Step 3: Start decoding
    trg_indexes = [TRG_vocab["<sos>"]]

    for _ in range(max_len):
        trg_tensor = torch.tensor([trg_indexes[-1]]).to(device)

        with torch.no_grad():
            output, hidden = model.decoder(trg_tensor, hidden, encoder_outputs)

        pred_token = output.argmax(1).item()
        trg_indexes.append(pred_token)

        if pred_token == TRG_vocab["<eos>"]:
            break

    # Step 4: Convert output tokens to words
    trg_tokens = [TRG_itos[i] for i in trg_indexes if i not in [TRG_vocab["<sos>"], TRG_vocab["<eos>"], TRG_vocab["<pad>"]]]
    return " ".join(trg_tokens)

# Example Usage
test_sentences = [
    "my name is Pegah",
    "life is a fleeting dream"
]

for sent in test_sentences:
    translation = translate_sentence(model, sent, SRC_vocab, TRG_vocab, SRC_itos, TRG_itos, DEVICE)
    print(f"EN: {sent} ➔ FA: {translation}")
````

After training, when we use the model for translation inference, we observe that it still produces incorrect or memorized translations for new sentences — for example, translating "my name is Pegah" as "اسم من جان است". This happens because the model was trained on a very small dataset and mainly memorized fixed sentence patterns rather than learning to generalize across unseen inputs. Additionally, while the encoder benefits from pre-trained English GloVe embeddings, the decoder side, which generates Persian sentences, is trained entirely from scratch without any pre-trained knowledge of Persian. As a result, the model can only reproduce translations it has seen during training and struggles to adapt to new names or phrases. In short, although the English input side has semantic richness thanks to GloVe, the Persian output side relies solely on the limited training data, which restricts the model's overall generalization and fluency in real-world translation tasks.

````python
# Attention Visualization

# Step 2: Import required libraries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import arabic_reshaper
from bidi.algorithm import get_display
import torch

# Step 3: Define utility function to reshape Persian tokens
def reshape_persian_token(token):
    reshaped_text = arabic_reshaper.reshape(token)
    bidi_text = get_display(reshaped_text)
    return bidi_text

# Simple tokenizer
def tokenize(text):
    return text.lower().strip().split()

# Step 4: Define the final attention visualization function
def visualize_attention_manual(model, src_sentence, tgt_sentence, SRC_vocab, TRG_vocab, SRC_itos, TRG_itos, device):
    model.eval()

    # Tokenize the input sentence
    src_tokens = ["<sos>"] + tokenize(src_sentence) + ["<eos>"]
    src_indexes = [SRC_vocab.get(token, SRC_vocab.get("<pad>", 0)) for token in src_tokens]
    src_tensor = torch.tensor(src_indexes).unsqueeze(1).to(device)  # (seq_len, batch_size)

    with torch.no_grad():
        encoder_outputs, hidden = model.encoder(src_tensor)

    trg_indexes = [TRG_vocab["<sos>"]]
    attentions = []

    for _ in range(50):  # Allow longer target sentences
        trg_tensor = torch.tensor([trg_indexes[-1]]).to(device)
        with torch.no_grad():
            output, hidden = model.decoder(trg_tensor, hidden, encoder_outputs)

        pred_token = output.argmax(1).item()
        trg_indexes.append(pred_token)

        if pred_token == TRG_vocab["<eos>"]:
            break

        attn_weights = model.decoder.attention(hidden, encoder_outputs)
        attn_weights = attn_weights.squeeze(1).detach().cpu().numpy()
        attentions.append(attn_weights)

    if len(attentions) == 0:
        print("No attentions to plot!")
        return

    attn_matrix = np.stack(attentions)

    # Use the provided target sentence (Persian)
    tgt_tokens = tokenize(tgt_sentence)

    # Reshape Persian tokens for correct display
    tgt_tokens_reshaped = [reshape_persian_token(token) for token in tgt_tokens]

    # Step 5: Plot
    plt.figure(figsize=(12, 8))
    sns.heatmap(attn_matrix[::-1],  # Flip vertically
                xticklabels=src_tokens,
                yticklabels=tgt_tokens_reshaped[::-1],  # Flip Persian labels too
                cmap="coolwarm")

    plt.xlabel("Source (English)")
    plt.ylabel("Target (Persian)")
    plt.title("Attention Heatmap: English ➔ Persian (Manual Target Text)")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Step 6: Try it out
visualize_attention_manual(
    model=model,
    src_sentence="i am at home",
    tgt_sentence="من در خانه هستم",
    SRC_vocab=SRC_vocab,
    TRG_vocab=TRG_vocab,
    SRC_itos=SRC_itos,
    TRG_itos=TRG_itos,
    device=DEVICE
)
````
