# Understanding Clustering as an Unsupervised Learning Technique

## 1. What is Clustering?
Clustering is a fundamental technique in unsupervised learning, where the goal is to group similar data points together without predefined labels. Unlike supervised learning, where models learn from labeled data, clustering identifies inherent structures in data based on similarity or density.

## 2. Why Use Clustering?

### Clustering helps in:

- Identifying hidden patterns or structures in data.
- Data segmentation and organization.
- Reducing data complexity by grouping similar instances.
- Serving as a preprocessing step for other machine learning algorithms (e.g., anomaly detection).

## 3. Key Concepts in Clustering

### Before diving into specific algorithms, it's essential to understand key concepts:

- Clusters: Groups of similar data points.
- Centroids: Representative points (or means) of clusters.
- Distance Measures: Methods to evaluate similarity between points (e.g., Euclidean distance, cosine similarity).
- Cluster Density: Defines the tightness of data points within a cluster.
- Cluster Shape: Some algorithms assume clusters are spherical, while others allow arbitrary shapes.

## 4. Types of Clustering Approaches

### Clustering algorithms can be broadly classified into three types:

- Centroid-based (e.g., K-Means): Assigns data points to a predefined number of clusters using centroids.
- Hierarchical (e.g., Agglomerative, Divisive): Forms a tree-like structure of clusters.
- Density-based (e.g., DBSCAN): Identifies clusters based on data density.

## 5. Applications of Clustering

- Customer Segmentation (Marketing)
- Anomaly Detection (Fraud detection, Cybersecurity)
- Image Segmentation (Computer Vision)
- Genetic Data Analysis (Bioinformatics)
- Recommendation Systems (E-commerce, Content streaming)

## Deep Dive into Clustering Concepts

When discussing clustering, it's essential to understand the fundamental components that define a clustering algorithm's behavior. Below, we break down key concepts in clustering along with mathematical formulations where necessary.

### 1. Clusters: Groups of Similar Data Points
Clusters are groups of similar data points formed based on some measure of similarity or density. Each cluster contains points that are more similar to each other than to points in other clusters.

The number and shape of clusters depend on the algorithm used (e.g., K-Means assumes spherical clusters, while DBSCAN allows arbitrary shapes).

### 2. Centroids: Representative Points of Clusters
Centroids represent the center of a cluster and are used in centroid-based clustering algorithms like K-Means. The centroid of a cluster is typically calculated as the mean of all points in the cluster:

$$
C_k = \frac{1}{N_k} \sum_{i=1}^{N_k} x_i
$$

where:
- $C_k$ is the centroid of the $k$-th cluster,
- $N_k$ is the number of points in cluster $k$,
- $x_i$ represents individual data points in cluster $k$.

Centroid-based methods assume clusters are convex and isotropic (same in all directions), which may not always be true in real-world data.

### 3. Distance Measures: Evaluating Similarity
Clustering requires a way to measure similarity between points, commonly using distance metrics. Some widely used distance measures include:

### a) Euclidean Distance (Most Common)
Measures the straight-line distance between two points:





euclidean_distance.png



where:
- $x$ and $y$ are two points in an $n$-dimensional space.

### b) Manhattan Distance (L1 Distance)
Measures distance along grid-like paths (sum of absolute differences):

$$
d(x, y) = \sum_{i=1}^{n} |x_i - y_i|
$$

### c) Cosine Similarity (for Text & High-Dimensional Data)
Measures the cosine of the angle between two vectors:

$$
\text{similarity}(x, y) = \frac{x \cdot y}{\|x\| \|y\|}
$$

where:
- $x \cdot y$ is the dot product of vectors $x$ and $y$,
- $\|x\|$ and $\|y\|$ are the magnitudes (norms) of the vectors.

Choosing the right distance metric depends on the dataset’s nature—Euclidean works well for compact clusters, while Cosine is better for high-dimensional data like text.

### 4. Cluster Density: How Tight Are the Clusters?
Cluster density refers to how closely packed the points are within a cluster. Density-based clustering algorithms like DBSCAN (Density-Based Spatial Clustering of Applications with Noise) use density thresholds to define clusters.

A region is considered dense if it has at least MinPts (minimum points) within a distance of $\epsilon$ (epsilon):

- **Core Point**: A point with at least MinPts neighbors within distance **$\epsilon$**.
- **Border Point**: A point with fewer than MinPts neighbors but within the **$\epsilon$** radius of a core point.
- **Noise Point**: A point that doesn’t belong to any cluster.

DBSCAN is useful for detecting clusters of varying shapes and ignoring outliers.

### 5. Cluster Shape: Spherical vs. Arbitrary
Different clustering algorithms assume different cluster shapes:

1. Spherical Clusters (K-Means)
- Works well when clusters are approximately globular (spherical) and well-separated.
- Assumes equal variance in all directions (isotropic clusters).
Example: Gaussian-distributed clusters with similar density and size.
2. Hierarchical Clustering
- Can form clusters of different sizes and shapes (not restricted to spherical clusters).
- Can capture nested hierarchical structures in data.
The results depend on the linkage method and distance metric used.
3. Density-Based Clustering (DBSCAN)
- Handles arbitrarily shaped clusters, including elongated and non-convex structures.
- Does not assume a spherical structure like K-Means.
Can detect outliers as noise points, making it useful for datasets with noise.

The best algorithm depends on the dataset's distribution and structure.

# **K-Means Clustering (Centroid-Based)**

K-Means is a centroid-based clustering algorithm that partitions data into K clusters by iteratively refining cluster centers. It assumes clusters are spherical and equally sized, making it suitable for well-separated, dense clusters.

## How K-Means Works (Step-by-Step)
1. Initialize: Choose K clusters and randomly initialize their centroids.
2. Assign: Each data point is assigned to the nearest centroid.
3. Update: Recompute the centroids by averaging all points in each cluster.
4. Repeat until centroids no longer move significantly (convergence).

## Mathematical Formulation
K-Means aims to minimize the intra-cluster variance, measured as:

$$
J = \sum_{i=1}^{K} \sum_{x \in C_i} ||x - \mu_i||^2
$$

where:
- $K$ is the number of clusters.
- $C_i$ is the set of points in cluster $i$.
- $\mu_i$ is the centroid of cluster $i$.
- $||x - \mu_i||^2$ is the squared Euclidean distance between a point $x$ and its centroid.
- $J$ (inertia) is the sum of squared distances between all points and their assigned cluster centroids, measuring how well the clusters fit the data.

## Choosing the Right Value of K
A common method to find the optimal number of clusters is the Elbow Method, which:
- Computes inertia (sum of squared distances) for different values of $K$.
- Plots inertia vs. K and looks for the "elbow" where inertia stops decreasing significantly.

## Convex vs. Non-Convex Shapes:

- Convex Shape: A shape where any two points inside can be connected by a straight line that remains completely inside the shape.
Examples: Circle, square, rectangle, oval.

- Non-Convex (Concave) Shape: A shape that has at least one indentation, meaning a straight line between two points can go outside the shape.
Examples: Star, crescent, L-shaped region.

**Are Shapes Only Convex or Non-Convex?**

Yes! There is no "in-between" because convexity is a strict mathematical property:

- A shape is convex if all straight-line segments between two points inside the shape remain inside.
- A shape is non-convex if at least one such segment extends outside the shape.
However, in practical applications, some shapes might be almost convex, meaning they have very slight indentations but still fall under the non-convex category.

**In Clustering & Machine Learning:**
- A convex cluster looks like a compact, blob-like shape (e.g., circles, ellipses).
- A non-convex cluster can have irregular, elongated, or complex shapes (e.g., moons, spirals).
- K-Means works well with convex clusters but struggles with non-convex clusters (like two interlocking spirals).
- DBSCAN or Spectral Clustering is better for detecting non-convex clusters.
- Some clustering methods (like Spectral Clustering) transform non-convex data into a higher-dimensional space where clusters appear more convex.

Thus, while convex and non-convex shapes are distinct categories, in real-world clustering, we sometimes transform non-convex shapes to make them behave more like convex ones.

## Advantages & Limitations of K-Means Clustering
### **Advantages**
- Simple and easy to implement.
- Scales well to large datasets.
- Works well when clusters are well-separated and compact.

### **Limitations**
- Requires specifying \(K\) beforehand.
- Sensitive to outliers and initial centroid placement.
- Assumes clusters are convex and have similar variance.
- Can struggle with non-convex or varying density clusters.

## When to Use K-Means?
### **Use K-Means if:**
- Clusters are well-separated and compact.
- You need a fast, scalable clustering algorithm.
- You can determine \(K\) (e.g., using the Elbow Method).

### **Avoid K-Means if:**
- Clusters are irregularly shaped or vary in density (*use DBSCAN instead*).
- Your data contains many outliers (*use robust clustering methods*).
- Clusters overlap significantly or data contains categorical variables (*use k-modes or hierarchical clustering instead*).

### K-Means Clustering & Finding Optimal K
In this section, we will:
- Generate synthetic data with three natural clusters.
- Apply K-Means clustering to identify these clusters.
- Visualize the clustered data along with the centroids.
- Use the Elbow Method to determine the optimal number of clusters (K).

By the end, you will understand how K-Means partitions data and how to select the best K value using inertia.


````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate synthetic data with 3 clusters
# make_blobs creates a dataset with 300 samples, distributed around 3 centers.
# cluster_std defines the spread of the clusters, and random_state ensures reproducibility.
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# Apply K-Means clustering
# n_clusters=3 specifies that we want to find 3 clusters.
# random_state ensures consistent results across runs.
# n_init=10 means the algorithm runs 10 times with different centroid seeds and picks the best one.
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X)

# Get cluster labels and centroids
# labels_ gives the cluster assignments for each data point.
# cluster_centers_ gives the final centroid positions.
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Plot clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolor='k', alpha=0.7)  # Plot data points colored by cluster
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')  # Mark centroids
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("K-Means Clustering Visualization")
plt.legend()
plt.show()

# Finding the optimal number of clusters using the Elbow Method
inertia = []  # List to store inertia values for different k
K_range = range(1, 10)  # Testing k values from 1 to 9

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)  # Train K-Means with different k values
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)  # Store the sum of squared distances (inertia) for each k

# Plot the elbow method graph
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o', linestyle='--')  # Plot inertia vs. number of clusters
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia (Sum of Squared Distances)")
plt.title("Elbow Method for Optimal K")
plt.show()
````

### How cluster_std Affects Clusters:
- Low cluster_std (e.g., 0.2) → Clusters are tight and well-separated.
- High cluster_std (e.g., 2.0) → Clusters are spread out and may overlap.

So, cluster_std=1.0 means each cluster’s points are normally distributed around the center with a standard deviation of 1.0.

### 🍷 K-Means Clustering on UCI Wine Dataset

Now, we use the **UCI Wine Dataset**, which contains **13 chemical properties** of different wine samples, along with a **Class label** identifying the type of wine.

This dataset is widely used for clustering and classification tasks.  

In this section, we will:

✔ Load and preprocess real-world data from UCI Machine Learning Repository  
✔ Apply K-Means clustering to find natural groupings in the dataset  
✔ Use the Elbow Method to determine the optimal number of clusters  
✔ Visualize the clustering results using PCA  


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# -------------------------------
# Step 1: Load the Wine Dataset
# -------------------------------

# UCI Wine dataset URL
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"

# Column names based on dataset description
columns = [
    "Class", "Alcohol", "Malic_Acid", "Ash", "Alcalinity_of_Ash", "Magnesium",
    "Total_Phenols", "Flavanoids", "Nonflavanoid_Phenols", "Proanthocyanins",
    "Color_Intensity", "Hue", "OD280/OD315_of_Diluted_Wines", "Proline"
]

# Load dataset into a DataFrame
df = pd.read_csv(url, header=None, names=columns)

# Display first few rows
print(df.head())

# -------------------------------
# Step 2: Preprocess the Data
# -------------------------------

# Remove the "Class" column since K-Means is an unsupervised algorithm
X = df.drop(columns=["Class"])

# Standardize the dataset (K-Means is sensitive to scale)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -------------------------------
# Step 3: Find Optimal K using Elbow Method
# -------------------------------

inertia = []  # Store inertia values for different K values
K_range = range(1, 11)  # Test K values from 1 to 10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)  # Sum of squared distances to nearest cluster center

# Plot the Elbow Method graph
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o', linestyle='--')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia (Sum of Squared Distances)")
plt.title("Elbow Method for Optimal K (Wine Dataset)")
plt.show()

# -------------------------------
# Step 4: Apply K-Means Clustering
# -------------------------------

# Based on the elbow method graph, we choose K=3
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)

# Fit the model and get cluster labels
labels = kmeans.fit_predict(X_scaled)

# -------------------------------
# Step 5: Reduce Dimensions using PCA
# -------------------------------

# PCA reduces high-dimensional data to 2D for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)  # Transform the dataset

# -------------------------------
# Step 6: Transform Cluster Centers for Proper Visualization
# -------------------------------

# IMPORTANT: K-Means centroids are in the original high-dimensional space.
# We need to project them into the PCA-transformed space so they align correctly in the plot.
centers_pca = pca.transform(kmeans.cluster_centers_)

# -------------------------------
# Step 7: Visualize Clusters
# -------------------------------

plt.figure(figsize=(8, 6))

# Scatter plot of the data points, colored by cluster labels
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', edgecolor='k', alpha=0.7)

# Scatter plot for cluster centroids (corrected projection)
plt.scatter(centers_pca[:, 0], centers_pca[:, 1], c='red', marker='X', s=200, label='Centroids')

# Labels and title
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title(f"K-Means Clustering on UCI Wine Dataset (K={optimal_k})")
plt.legend()
plt.show()
````

- This visualization represents K-Means clustering applied to the UCI Wine Dataset, reduced to two principal components (PCs) using PCA. Each data point is colored according to its assigned cluster, while the red "X" markers represent the cluster centroids.


- By default, K-Means centroids are computed in the original high-dimensional space (13-dimensional feature space in this case). However, since we reduced the data to 2D using PCA, plotting the centroids directly from the original space would result in incorrect placement.

- To ensure proper alignment, we transformed the centroids using PCA before plotting them. This step correctly projects them into the same 2D space as the clustered data points, resulting in an accurate and meaningful visualization. Without this transformation, the centroids might appear misplaced and not truly represent the center of their respective clusters in the PCA-reduced space.

### 📊 Interpreting the Results

The Elbow Method plot suggests that K=3 is the optimal number of clusters, which aligns with the fact that the Wine dataset originally contains three wine classes. After applying K-Means clustering, we visualized the results using PCA to reduce the 13-dimensional feature space into 2D. The scatter plot shows how K-Means grouped the wine samples into three distinct clusters, with centroids marked in red Xs. While the clusters are well-separated, there may still be some overlap due to the natural similarities in chemical composition among certain wine types. This demonstrates how K-Means can uncover hidden patterns in data, even without knowing the original class labels.


# **Hierarchical Clustering: A Tree-Based Approach**
Hierarchical Clustering is a tree-based clustering technique that groups data points into clusters by building a hierarchy of nested clusters. Unlike K-Means, it doesn’t require specifying the number of clusters beforehand. Instead, it creates a dendrogram (a tree-like diagram) that helps determine the optimal number of clusters.


## Types of Hierarchical Clustering

There are two main approaches:

### Agglomerative Clustering (Bottom-Up)
- Starts with each data point as an individual cluster.
- Iteratively merges the closest clusters until only one remains.
- The most common approach in hierarchical clustering.

### Divisive Clustering (Top-Down)
- Starts with all data points in one large cluster.
- Recursively splits clusters into smaller groups until each point is its own cluster.
- Less common than agglomerative clustering.


Agglomerative clustering is more widely used because it is computationally efficient compared to divisive clustering.



````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram

# Generate synthetic data
X, _ = make_blobs(n_samples=50, centers=3, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ---------------------- Agglomerative Clustering (Bottom-Up) ---------------------- #

# Plot Dendrogram for Agglomerative Clustering
plt.figure(figsize=(6, 3))
sch.dendrogram(sch.linkage(X_scaled, method="ward"))
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.title("Agglomerative Clustering Dendrogram (Bottom-Up Approach)")
plt.show()

# Apply Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=3, metric="euclidean", linkage="ward")
labels_agg = agg_clustering.fit_predict(X_scaled)

# Scatter plot of clusters
plt.figure(figsize=(6, 3))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels_agg, cmap='viridis', edgecolor='k', alpha=0.7)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Agglomerative Clustering (Hierarchical Bottom-Up)")
plt.show()

# ---------------------- Divisive Clustering (Top-Down) ---------------------- #

# Compute the linkage matrix for Divisive Clustering
linkage_matrix = linkage(X_scaled, method="ward")

# Plot the dendrogram for Divisive Clustering (Simulated by Flipping)
plt.figure(figsize=(6, 3))
dendrogram(linkage_matrix, truncate_mode="lastp", p=10, show_leaf_counts=True)
plt.gca().invert_yaxis()  # Flip the dendrogram to simulate top-down splitting
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.title("Divisive Clustering Dendrogram (Top-Down Approach)")
plt.show()
````

## How Does Agglomerative Clustering Work?
The process follows these steps:

1. Compute the Distance Matrix  
   - Calculate the pairwise distance between all data points.  
   - Common distance metrics: Euclidean, Manhattan, Cosine Similarity.  

2. Merge Closest Clusters  
   - Identify the two closest clusters and merge them.  

3. Repeat Until One Cluster Remains  
   - This creates a hierarchical structure represented by a dendrogram.  

## Linkage Methods: How Do We Measure Cluster Similarity?
When merging clusters in **hierarchical clustering**, we use a **linkage criterion** to determine which clusters should be combined. Each method relies on a **distance metric** to measure similarity.

### 1. Single Linkage (Minimum Distance)
- Merges clusters based on the **closest** points between them.
- Can form long, chain-like clusters (**suffers from chaining effect**).

$$
D(A, B) = \min \{ d(x, y) : x \in A, y \in B \}
$$

**Common Distance Metrics:** Euclidean, Manhattan, Cosine  


### 2. Complete Linkage (Maximum Distance)
- Merges clusters based on the **farthest** points between them.
- Produces compact clusters but may break natural groupings.

$$
D(A, B) = \max \{ d(x, y) : x \in A, y \in B \}
$$

**Common Distance Metrics:** Euclidean, Manhattan, Cosine  


### 3. Average Linkage
- Merges clusters based on the **average** distance between all points.
- Balances between single and complete linkage.

$$
D(A, B) = \frac{1}{|A| |B|} \sum_{x \in A} \sum_{y \in B} d(x, y)
$$

**Common Distance Metrics:** Euclidean, Manhattan, Cosine  

### 4. Ward’s Method (Minimizing Variance)
- Minimizes the increase in variance when merging clusters.
- Tends to form compact, spherical clusters (**similar to K-Means**).

$$
D(A, B) = \sum_{i \in A \cup B} (x_i - \mu_{AB})^2 - \sum_{i \in A} (x_i - \mu_A)^2 - \sum_{i \in B} (x_i - \mu_B)^2
$$

**Common Distance Metric:** Euclidean only (since it relies on variance calculations).  

⭐ Ward’s Method is widely used as it creates compact, well-separated clusters.

### **Key Takeaways**
- Linkage defines how we measure the "distance" between clusters.
- Distance metrics define how we measure the "distance" between points.
- Different linkage methods can produce different clustering results, even with the same dataset and distance metric.

## 🧩 Approximating Divisive Clustering with **Spectral Clustering**

In the code below, **Spectral Clustering** is used as an approximation for Divisive Clustering because traditional divisive hierarchical clustering (DIANA) is not implemented in scikit-learn. Unlike agglomerative clustering (bottom-up), divisive clustering follows a top-down approach, where all points start in one cluster and are recursively split into smaller subgroups.

### Why Spectral Clustering?  
Although **Spectral Clustering is NOT a true divisive clustering method**, it shares some conceptual similarities:

- Graph-Based Approach: Instead of forming clusters using pairwise distances, Spectral Clustering represents data as a similarity graph.
- Global Partitioning: It identifies global structures first and partitions the dataset into meaningful subgroups, similar to divisive clustering.
- Eigenvalue Decomposition: It transforms the dataset into a lower-dimensional space using the Laplacian of the similarity matrix, where clusters are more separable.
- Assigning Clusters Using K-Means: After transforming the data, K-Means is applied to assign cluster labels, ensuring a structured and stable split.


### Key Differences: Spectral Clustering vs. True Divisive Clustering  
| Feature | Spectral Clustering | True Divisive Clustering (DIANA) |
|---------|---------------------|--------------------------------|
| **Approach** | Graph-based partitioning | Recursive hierarchical splitting |
| **Distance Metric** | Uses a similarity graph (e.g., nearest neighbors) | Based on pairwise distance (e.g., Euclidean) |
| **Computational Efficiency** | More scalable for large datasets | Computationally expensive |
| **Clustering Strategy** | Embeds data in a new space before clustering | Recursively splits original feature space |
| **Hierarchical?** | ❌ No, but can approximate top-down clustering | ✅ Yes, explicitly hierarchical |




````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import SpectralClustering

# Step 1: Load the dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
columns = [
    "Class", "Alcohol", "Malic_Acid", "Ash", "Alcalinity_of_Ash", "Magnesium",
    "Total_Phenols", "Flavanoids", "Nonflavanoid_Phenols", "Proanthocyanins",
    "Color_Intensity", "Hue", "OD280/OD315_of_Diluted_Wines", "Proline"
]
df = pd.read_csv(url, header=None, names=columns)

# Step 2: Preprocess the Data
X = df.drop(columns=["Class"])  # Remove labels (unsupervised learning)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # Normalize features for better clustering

# Step 3: Apply Divisive Clustering using Spectral Clustering
optimal_k = 3  # Choose the number of clusters (based on dendrogram analysis)
divisive_clustering = SpectralClustering(n_clusters=optimal_k, affinity='nearest_neighbors', assign_labels='kmeans', random_state=42)
labels = divisive_clustering.fit_predict(X_scaled)

# Step 4: Visualize Clusters Using PCA (2D Projection)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', edgecolor='k', alpha=0.7)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title(f"Divisive Clustering (Spectral Clustering) on UCI Wine Dataset (K={optimal_k})")
plt.show()
````

## Introduction to Eigenvalues, Eigenvectors, and the Laplacian Matrix

### Eigenvalues and Eigenvectors
Eigenvalues and eigenvectors are fundamental concepts in linear algebra used in various applications, including Principal Component Analysis (PCA) and Spectral Clustering.

- **Eigenvector ($v$)**: A nonzero vector that changes only in scale (not direction) when multiplied by a matrix.
- **Eigenvalue ($\lambda$)**: A scalar that represents how much the eigenvector is stretched or shrunk.

Mathematically, for a square matrix $A$:
$$ A v = \lambda v $$
where:
- $A$ is an $n \times n$ matrix,
- $v$ is the eigenvector,
- $\lambda$ is the corresponding eigenvalue.

### Laplacian Matrix ($L$)
The Laplacian matrix is a key concept in graph-based clustering methods such as Spectral Clustering. It captures the structure of a graph and is used to find meaningful partitions.

#### Definition of the Graph Laplacian
Given a graph represented by an **adjacency matrix** $A$ and a **degree matrix** $D$:
$$ L = D - A $$
where:
- $D$ is a diagonal matrix where each entry $D_{ii}$ is the sum of the degrees of node $i$,
- $A$ is the adjacency matrix of the graph.

#### Normalized Laplacian Matrix
For spectral clustering, a normalized version of the Laplacian matrix is often used to improve numerical stability:
$$ L_{\text{sym}} = D^{-1/2} L D^{-1/2} = I - D^{-1/2} A D^{-1/2} $$
where:
- $L_{\text{sym}}$ is the symmetric normalized Laplacian,
- $I$ is the identity matrix,
- $D^{-1/2}$ scales the Laplacian to balance node degrees.

This formulation allows for efficient clustering by computing the **eigenvalues and eigenvectors** of $L_{\text{sym}}$, which reveal the graph’s structure.



# **Density-Based Clustering (DBSCAN)**

## What is DBSCAN?
Density-Based Spatial Clustering of Applications with Noise (**DBSCAN**) is a **density-based** clustering algorithm that identifies clusters as dense regions of points separated by low-density areas. Unlike K-Means and Hierarchical Clustering:
- DBSCAN does not require specifying the number of clusters (K) beforehand.
- It can identify clusters of arbitrary shapes (not just spherical ones).
- It can detect outliers as noise.

## How DBSCAN Works
DBSCAN groups points based on density connectivity, controlled by two key **hyperparameters**:

- **$\varepsilon$ (epsilon)**: The maximum neighborhood radius—if two points are within this distance, they are considered neighbors.
- **MinPts**: The minimum number of points required within **$\varepsilon$** to form a dense region.

These two hyperparameters determine how clusters are formed and significantly impact DBSCAN's performance.

### Steps of DBSCAN:
1. For each point, find all neighboring points within distance **$\varepsilon$**.
2. Classify points as:
   - **Core Points**: Points with at least MinPts neighbors within **$\varepsilon$**.
   - **Border Points**: Points within $\varepsilon$ of a core point but with fewer than MinPts neighbors.
   - **Noise Points**: Points that do not belong to any cluster.
3. Expand clusters by recursively adding density-connected points.
4. Repeat until all points are classified.

## Mathematical Definition of Density Connectivity
A point **$p$** is directly density-reachable from a point **$q$** if:

$$
q \in N_{\varepsilon}(p) \quad \text{and} \quad |N_{\varepsilon}(p)| \geq \text{MinPts}
$$

where **$N_{\varepsilon}(p)$** is the set of points within distance **$\varepsilon$** from **$p$**.

A point **$p$** is density-reachable from **$q$** if there exists a chain of points $p_1, p_2, ..., p_n$ such that:

$$
 p_1 = q, \quad p_n = p, \quad \text{and} \quad p_{i+1} \text{ is directly density-reachable from } p_i.
$$

Two points **$p$** and **$q$** are density-connected if there exists a third point $o$ that can reach both **$p$** and **$q$** through density-reachability.

## Hyperparameters in DBSCAN:
### Understanding $\varepsilon$ (Epsilon)
- Controls how close points must be to be considered neighbors.
- **Small $\varepsilon$** → Many small clusters or too many outliers.
- **Large $\varepsilon$** → Merges clusters together.

### Understanding MinPts
- Controls how many neighbors a point must have to be considered a core point.
- **Low MinPts** → Can detect smaller clusters but may be too sensitive to noise.
- **High MinPts** → Forms more robust clusters but may miss smaller ones.

### Hyperparameter Tuning Tip
- Use a k-distance plot to find the best **$\varepsilon$**.
- A common rule: MinPts should be at least **2 × number of dimensions** in the dataset.

## Advantages of DBSCAN
- No need to specify K (number of clusters is automatically determined).
- Can identify clusters of arbitrary shapes (works well for non-spherical clusters).
- Detects outliers naturally (points not belonging to any cluster are labeled as noise).
- Performs well when clusters have varying densities, unlike K-Means.

## Limitations of DBSCAN
- Sensitive to $\varepsilon$: A bad choice can either break clusters apart or merge them incorrectly.
- Struggles in high-dimensional data: Distance calculations become less meaningful in many dimensions.
- Fails when clusters vary significantly in density: A single **$\varepsilon$** may not work for all clusters.

## When to Use DBSCAN?
**Use DBSCAN if:**
- Clusters are irregularly shaped (e.g., spirals, blobs, and non-convex regions).
- You don’t know the number of clusters (K) in advance.
- You need to detect outliers as part of clustering.

**Avoid DBSCAN if:**
- The dataset is high-dimensional (distances become less reliable).
- The dataset has clusters with very different densities (DBSCAN struggles with this).
- Choosing **$\varepsilon$** is too difficult due to varying densities.

### 💡 Are $\varepsilon$ and MinPts Hyperparameters?
Yes! Both $\varepsilon$ and MinPts are hyperparameters because they control how DBSCAN defines dense regions and expands clusters. Their values must be tuned based on the dataset, and different settings can lead to different clustering results.


### DBSCAN Clustering on UCI Wine Dataset
In this section, we apply Density-Based Spatial Clustering of Applications with Noise (DBSCAN) to the UCI Wine Dataset, which contains 13 chemical properties of different wine samples.

### What This Code Does
1. Loads and Preprocesses the UCI Wine Dataset  
   - Standardizes features to ensure fair distance calculations.  

2. Applies DBSCAN Clustering  
   - Uses Euclidean distance and optimized `eps=1.8, min_samples=4`.  
   - Detects clusters and noise points (outliers).

3. Visualizes Clustering Results Using PCA  
   - Reduces high-dimensional data to 2D for visualization.  
   - Clusters are color-coded, while noise points appear as red "X" markers.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Step 1: Load the dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
columns = [
    "Class", "Alcohol", "Malic_Acid", "Ash", "Alcalinity_of_Ash", "Magnesium",
    "Total_Phenols", "Flavanoids", "Nonflavanoid_Phenols", "Proanthocyanins",
    "Color_Intensity", "Hue", "OD280/OD315_of_Diluted_Wines", "Proline"
]
df = pd.read_csv(url, header=None, names=columns)

# Step 2: Preprocess the Data
X = df.drop(columns=["Class"])  # Remove labels (unsupervised learning)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # Normalize features for better clustering

# Step 3: Apply DBSCAN with Euclidean Distance (Better Performance)
dbscan = DBSCAN(eps=1.8, min_samples=4, metric='euclidean')  # Go back to Euclidean
labels = dbscan.fit_predict(X_scaled)

# Step 4: Visualize Clusters Using PCA (2D Projection)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8, 6))
unique_labels = set(labels)

# Define distinct colors for clusters (Fully fixed)
colors = plt.colormaps.get_cmap("tab10").colors  # Get colors correctly

for label, color in zip(unique_labels, colors[:len(unique_labels)]):  # Use only required colors
    if label == -1:
        plt.scatter(X_pca[labels == label, 0], X_pca[labels == label, 1],
                    c='red', marker='x', s=50, label="Noise", alpha=0.7)  # Make noise smaller & red
    else:
        plt.scatter(X_pca[labels == label, 0], X_pca[labels == label, 1],
                    color=color, edgecolor='k', alpha=0.7, label=f"Cluster {label}")

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("DBSCAN Clustering on UCI Wine Dataset (Optimized)")
plt.legend()
plt.show()
````

### Why This Clustering is Useful

✔ Better at detecting non-spherical clusters compared to K-Means.  
✔ Identifies anomalies (noise points), which can be useful for detecting unusual wine samples.  



**🔍 Why Apply DBSCAN Before PCA?**

In this implementation, DBSCAN is applied to the full-dimensional, standardized dataset before PCA is used for visualization. This ensures that clustering is based on the true feature space rather than a reduced-dimensional approximation.

**Key Reasons for This Approach:**

- DBSCAN is density-based: It relies on distances between points to determine dense regions, and reducing dimensions with PCA before clustering can distort these distances, affecting cluster formation.
- PCA is for visualization, not clustering: PCA reduces the dataset to two dimensions for easy visualization, but this simplification can merge or distort clusters that were separable in higher dimensions.
- Correct Noise Detection: Running DBSCAN in the original feature space ensures that noise points are correctly identified based on actual density variations. If PCA is applied first, the noise structure may be lost, leading to missing or misclassified noise points.

By following this approach, we preserve the true clustering structure while using PCA only as a tool to project the results into a human-interpretable 2D space. This maintains the accuracy of DBSCAN while making the clusters easier to visualize.

- DBSCAN is designed to work on original feature space: Since DBSCAN detects clusters based on density, reducing the number of dimensions before clustering may eliminate meaningful density patterns.

- Can lead to excessive noise points (-1 labels): When PCA reduces high-dimensional data into 2D, points that were dense in high dimensions may become sparse, increasing the number of points classified as noise.


# Clustering Performance Evaluation Metrics

## 1. Internal Evaluation Metrics (No Ground Truth Needed)
Internal metrics evaluate clustering quality **without labels** by measuring **cohesion** (how compact clusters are) and **separation** (how distinct clusters are).

### a) Silhouette Score
Silhouette Score measures how well a data point is clustered by comparing its **intra-cluster distance** (cohesion) and **nearest-cluster distance** (separation).

$$
S(i) = \frac{b(i) - a(i)}{\max(a(i), b(i))}
$$

where:
- \( a(i) \) = average distance between **point \(i\)** and all other points in the **same cluster**.
- \( b(i) \) = average distance between **point \(i\)** and the **nearest neighboring cluster**.

Score Ranges:
- **+1** $ \rightarrow $ Well-clustered.
- **0** $ \rightarrow $ Overlapping clusters.
- **-1** $ \rightarrow $ Poor clustering.


### b) Davies-Bouldin Index (DBI)
Measures the **compactness** and **separation** of clusters.

$$
DBI = \frac{1}{N} \sum_{i=1}^{N} \max_{j \neq i} \left( \frac{\sigma_i + \sigma_j}{d_{ij}} \right)
$$

where:
- $ \sigma_i $ = average distance from points in **cluster $ i $** to its centroid.
- $ d_{ij} $ = distance between centroids of **clusters $ i $ and $ j $**.


Lower DBI is better (indicates compact and well-separated clusters).


### c) Calinski-Harabasz Index (CH Score)
Measures the ratio of **between-cluster variance to within-cluster variance**.

$$
CH = \frac{Tr_B / (K - 1)}{Tr_W / (N - K)}
$$

where:
- $ Tr_B $ = variance **between clusters**.
- $ Tr_W $ = variance **within clusters**.
- $ K $ = number of clusters.
- $ N $ = total points.


Higher CH Index is better (indicates well-separated clusters).

## 2. External Evaluation Metrics (When Ground Truth Labels Are Available)
Used **only when ground-truth labels exist** to measure how well clustering matches known classes.

### a) Adjusted Rand Index (ARI)
Measures how well clusters match true labels.

$$
ARI = \frac{\sum_{ij} \binom{n_{ij}}{2} - \left[ \sum_i \binom{a_i}{2} \sum_j \binom{b_j}{2} \right] / \binom{N}{2} }{\frac{1}{2} \left[ \sum_i \binom{a_i}{2} + \sum_j \binom{b_j}{2} \right] - \left[ \sum_i \binom{a_i}{2} \sum_j \binom{b_j}{2} \right] / \binom{N}{2} }
$$

Ranges from -1 to 1:
- **+1** $ \rightarrow $ Perfect clustering.
- **0** $ \rightarrow $ Random clustering.
- $ \mathbf{Negative} \rightarrow $ Worse than random.


### b) Normalized Mutual Information (NMI)
Measures the similarity between the clustering and ground truth labels based on **entropy**.

$$
NMI = \frac{2 I(Y, C)}{H(Y) + H(C)}
$$

where:
- $ I(Y, C) $ = **mutual information** between **true labels** ($ Y $) and **predicted clusters** ($ C $).
- $ H(Y) $ and $ H(C) $ = **entropy** of **true labels** and **predicted clusters**.


Ranges from $0$ to $1$:
- **1** $ \rightarrow $ Perfect clustering.
- **0** $ \rightarrow $ No correlation.



## 3. Stability-Based Evaluation (Cluster Reproducibility)

### a) Running K-Means Multiple Times
If clustering is **consistent**, results should be stable across different runs.

$$
\sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (S_i - \bar{S})^2
$$

where:
- $ S_i $ = Silhouette Score in the **$i^{th}$** run.
- $ \bar{S} $ = **Mean** Silhouette Score across runs.
- $ n $ = Number of runs.


Lower variance means more stable clustering.



## 4. Which Metric Works Best for Each Algorithm?

\begin{array}{|c|c|c|c|c|}
\hline
\textbf{Algorithm} & \textbf{Silhouette Score} & \textbf{Davies-Bouldin Index} & \textbf{ARI / NMI (If Labels Exist)} & \textbf{Stability} \\
\hline
\textbf{K-Means} & \checkmark \text{ Good} & \checkmark \text{ Good} & \checkmark \text{ Works} & \checkmark \text{ Run Multiple Times} \\
\hline
\textbf{Hierarchical} & \checkmark \text{ Good} & \checkmark \text{ Good} & \checkmark \text{ Works} & \checkmark \text{ Reproducible} \\
\hline
\textbf{DBSCAN} & \times \text{ Not Always Useful} & \checkmark \text{ Works} & \checkmark \text{ Works} & \checkmark \text{ Parameter Tuning Required} \\
\hline
\end{array}


## 5. Key Takeaways
✔ **Silhouette Score** → Best for K-Means & Hierarchical Clustering.

✔ **Davies-Bouldin Index & Calinski-Harabasz** → Measure cluster compactness & separation.

✔ **ARI & NMI** → Compare clustering with true labels (if available).

✔ **Stability Test** → Ensures clustering is reproducible across runs.

💡 **For DBSCAN**, focus on Davies-Bouldin Index and ARI/NMI (if labels exist).


### Final Notes
Evaluating clustering requires choosing the right metric for the algorithm. Internal metrics work without labels, while external metrics require ground truth. Stability tests help check reproducibility, especially for K-Means.

**Choose the right evaluation metric for your clustering problem!**

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score, adjusted_rand_score, normalized_mutual_info_score

# Step 1: Load the UCI Breast Cancer Wisconsin Dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
columns = ["ID", "Diagnosis"] + [f"Feature_{i}" for i in range(1, 31)]
df = pd.read_csv(url, header=None, names=columns)

# Step 2: Preprocess Data
X = df.drop(columns=["ID", "Diagnosis"])  # Remove ID and Diagnosis (labels)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # Normalize features

# Step 3: Apply Clustering Algorithms

# K-Means Clustering
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_scaled)

# Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=2, linkage="ward")
hierarchical_labels = hierarchical.fit_predict(X_scaled)

# DBSCAN Clustering
dbscan = DBSCAN(eps=1.8, min_samples=5, metric='euclidean')
dbscan_labels = dbscan.fit_predict(X_scaled)

# Step 4: Evaluate Clustering Performance

# Function to compute clustering evaluation metrics
def evaluate_clustering(X, labels, true_labels=None):
    if len(set(labels)) > 1:  # Only compute metrics if more than 1 cluster found
        silhouette = silhouette_score(X, labels)
        dbi = davies_bouldin_score(X, labels)
        chi = calinski_harabasz_score(X, labels)
        ari = adjusted_rand_score(true_labels, labels) if true_labels is not None else None
        nmi = normalized_mutual_info_score(true_labels, labels) if true_labels is not None else None
    else:
        silhouette, dbi, chi, ari, nmi = None, None, None, None, None
    return silhouette, dbi, chi, ari, nmi

# True labels (0 = Benign, 1 = Malignant)
true_labels = (df["Diagnosis"] == 'M').astype(int)  # Convert 'M' (Malignant) to 1, 'B' (Benign) to 0

# Compute metrics for each clustering method
metrics = {
    "K-Means": evaluate_clustering(X_scaled, kmeans_labels, true_labels),
    "Hierarchical": evaluate_clustering(X_scaled, hierarchical_labels, true_labels),
    "DBSCAN": evaluate_clustering(X_scaled, dbscan_labels, true_labels)
}

# Step 5: Display Clustering Results
results_df = pd.DataFrame.from_dict(metrics, orient='index', columns=[
    "Silhouette Score", "Davies-Bouldin Index", "Calinski-Harabasz Index",
    "Adjusted Rand Index", "Normalized Mutual Info"
])

# Replace None values with "N/A" for better readability
results_df = results_df.fillna("N/A")

# Print the results for quick inspection
print("\n=== Clustering Performance Metrics ===")
print(results_df)

# Display the results in a proper table format for Colab
from IPython.display import display
display(results_df)  # This will show a nice table in Google Colab


````
